## **Merg()**

```python
#Since all data share one column (store) I will merge them all start with merging train and feature first
merge_train_features = pd.merge(train, features, on=["Store", "Date"], how="left")
#And then merge (train and feature) with store
full_df = pd.merge(merge_train_features, stores, on="Store", how="left")
```

## **Groupby + Aggregation**

### **1.sum**

```python
Weekly_Sales_Dep=full_df.groupby('Dept')['Weekly_Sales'].sum()
Weekly_Sales_Dep
```

### **2. mean**

```python
Weekly_Sales_Dep=full_df.groupby('Dept')['Weekly_Sales'].mean()
Weekly_Sales_Dep
```

### **3. max**

```python
Weekly_Sales_Dep=full_df.groupby('Dept')['Weekly_Sales'].max()
Weekly_Sales_Dep
```

### **4. min**

```python
Weekly_Sales_Dep=full_df.groupby('Dept')['Weekly_Sales'].min()
Weekly_Sales_Dep
```

### **5. count**

```python
Weekly_Sales_Dep=full_df.groupby('Dept')['Weekly_Sales'].count()
Weekly_Sales_Dep
```

## **concat()**

```python
combine_train_features = pd.concat([train, features], axis=0, ignore_index=True)
combine_train_features.head(10)
```

## **join()**

```python
Weekly_Sales_per_Dept= train.groupby('Dept')['Weekly_Sales'].mean().reset_index()
Weekly_Sales_per_Dept.rename(columns={'Weekly_Sales': 'Avg_Weekly_Sales'}, inplace=True)
train_with_avg_sales = train.merge(Weekly_Sales_per_Dept, on='Dept', how='left')
train_with_avg_sales.head(10)
```
