```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y=np.sin(x)
y2=np.cos(x)
print("x:", x)
print("y:",y)
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
#To add a label, change the color of the graph, type of grapg line (we can use '---' as well)
plt.plot(x,y,label = 'sin(x)', color='red', linestyle=':')
plt.plot(x,y2,label = 'sin(x)', color='blue', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(26,6)) #First width and then length
plt.plot(x,y,label = 'sin(x)', color='red', linestyle=':')
plt.plot(x,y2,label = 'sin(x)', color='blue', linestyle='--')
#Adding labels to the x-axis and y-axis
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.figure(figsize=(20,10)) #First width and then length
plt.plot(x,y,label = 'sin(x)', color='red', linestyle=':')
plt.plot(x,y2,label = 'sin(x)', color='blue', linestyle='--')
#Adding labels to the x-axis and y-axis and changing font size
plt.xlabel("X-axis", fontsize =16)
plt.ylabel("Y-axis", fontsize = 16)
#To change the font size
plt.show()
```

```python
plt.figure(figsize=(20,10)) #First width and then length
plt.plot(x,y,label = 'sin(x)', color='red', linestyle=':')
plt.plot(x,y2,label = 'sin(x)', color='blue', linestyle='--')
#Adding labels to the x-axis and y-axis and changing font size
plt.xlabel("X-axis", fontsize =16)
plt.ylabel("Y-axis", fontsize = 16)
#To add grid
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,10)) #First width and then length
plt.plot(x,y,label = 'sin(x)', color='red', linestyle=':')
plt.plot(x,y2,label = 'sin(x)', color='blue', linestyle='--')
#Adding labels to the x-axis and y-axis and changing font size
plt.xlabel("X-axis", fontsize =16)
plt.ylabel("Y-axis", fontsize = 16)
#To add grid
plt.grid(True)
#To add title and changing it's size
plt.title ("Sin and Cos waves", fontsize =16)
#To print the labels
plt.legend()
plt.show()
```

```python
np.random.seed(42)
x_scatter = np.random.rand(50) *10
y_scatter = np.random.rand(50) *10 +x_scatter *0.5 + np.random.randn(50)

plt.figure(figsize=(15,8))
#Alpha make the dots / lines more transparent, edgecolors change the outline of the dots
plt.scatter(x_scatter, y_scatter, color='red', alpha=0.6, edgecolors ="black")
plt.title("Scatter Plot of X vs Y", fontsize=16)
plt.xlabel("X values", fontsize=16)
plt.ylabel("Y values", fontsize=16)
plt.legend()
plt.grid(True, linestyle = ':', alpha=0.6)
plt.show()
```

```python
categories = ['A','B','C','D','E']
values=[23,45,65,18,39]

plt.figure(figsize=(15,8))
plt.bar(categories,values , color=['skyblue','lightcoral','lightgreen','gold','plum'])
plt.title("Bar Plot of Categories Values", fontsize=16)
plt.xlabel("Categories" ,fontsize=16)
plt.ylabel("Values",fontsize=16)
plt.grid(True, alpha =0.4)
plt.legend()
plt.show()
```

```python
categories = ['A','B','C','D','E']
values=[23,45,65,18,39]

plt.figure(figsize=(15,8))
bars=plt.bar(categories,values , color=['skyblue','lightcoral','lightgreen','gold','plum'])
for bar in bars:
    yval=bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center' , va='bottom')
plt.title("Bar Plot of Categories Values", fontsize=16)
plt.xlabel("Categories" ,fontsize=16)
plt.ylabel("Values",fontsize=16)
plt.grid(True, alpha =0.4)
plt.legend()
plt.show()

```
