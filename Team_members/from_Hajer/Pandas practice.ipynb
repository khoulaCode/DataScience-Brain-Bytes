# **Groupby part**

### **1. Top 5 countries by average GDP**

```python
#Top 5 countries by average GDP
top_5_countries_gdp = df.groupby(" Country ")[" Gross Domestic Product (GDP) "].mean().sort_values(ascending=False).head(5)
top_5_countries_gdp
```

```python
#Plotting Top 5 countries by average GDP
top_5_countries_gdp = df.groupby(" Country ")[" Gross Domestic Product (GDP) "].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(8, 8))
categories = top_5_countries_gdp.index
values = top_5_countries_gdp.values
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.title("Top 5 Countries by Average GDP", fontsize=12, color='blue')
plt.ylabel("Average GDP (USD)", fontsize=12, color='blue')
plt.xlabel("Country", fontsize=12, color='blue')
plt.grid(True, alpha=0.4)
plt.show()
```

### **2.Countries with the Household consumption expenditure in the most recent year**

```python
# Countries with the Household consumption expenditure in the most recent year
Household_consumption_expenditure = df.groupby(" Country ")[ ' Household consumption expenditure (including Non-profit institutions serving households) '].max().sort_values(ascending=False).head(5)
Household_consumption_expenditure
```

```python
#Plotting Countries with the Household consumption expenditure in the most recent year
Household_consumption_expenditure = df.groupby(' Country ')[ ' Household consumption expenditure (including Non-profit institutions serving households) '].max().head(10)
plt.figure(figsize=(10, 10))
categories = Household_consumption_expenditure.index
values = Household_consumption_expenditure.values
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.title( "Household consumption expenditure (including Non-profit institutions serving households) ", fontsize=12, color='blue')
plt.ylabel("Household consumption expenditure", fontsize = 12, color='blue')
plt.xlabel("Country", fontsize=12, color='blue')
plt.grid(True, alpha=0.4)
plt.show()
```

### **3. Group by Country and calculate GDP per capita**

```python
# Group by Country and calculate:
# GDP per capita
df[' GDP Per capita GNI '] = df[' Gross Domestic Product (GDP) '] / df[' Per capita GNI ']
df[' GDP Per capita GNI ']
```

```python
mean_gdp_per_captia = df.groupby(' Country ')[' GDP Per capita GNI '].mean().sort_values(ascending=False).head(10)
mean_gdp_per_captia
```

```python
#Plotting Country GDP per capita
mean_gdp_per_captia = df.groupby(' Country ')[' GDP Per capita GNI '].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 10))
categories = mean_gdp_per_captia.index
values = mean_gdp_per_captia.values
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.title("Mean GDP per capita", fontsize=12, color='blue')
plt.ylabel("Average GDP (USD)", fontsize=12, color='blue')
plt.xlabel("Country", fontsize=12, color='blue')
plt.grid(True, alpha=0.4)
plt.show()
```

### **4. Max import of goods values**

```python
#Max import of goods values
Max_goods_values= df.groupby(' Country ')[' Imports of goods and services '].max()
Max_goods_values
```

```python
#Plotting Max import of goods values
Max_goods_values= df.groupby(' Country ')[' Exports of goods and services '].sum().head(10)
plt.figure(figsize=(8, 10))
categories = Max_goods_values.index
values = Max_goods_values.values
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.title("Total export overall years", fontsize=12, color='blue')
plt.ylabel("Exports of goods and services", fontsize=12, color='blue')
plt.xlabel("Country", fontsize=12, color='blue')
plt.grid(True, alpha=0.4)
plt.show()
```

### **5. Total exports over all years**

```python
# Total exports over all years
Total_exports_overall_years= df.groupby(' Country ')[' Exports of goods and services '].sum().head(10)
Total_exports_overall_years
```

```python
# Plotting Total exports over all years
Total_exports_overall_years= df.groupby(' Country ')[' Exports of goods and services '].sum().head(10)
plt.figure(figsize=(8, 10))
categories = Total_exports_overall_years.index
values = Total_exports_overall_years.values
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
plt.title("Total export overall years", fontsize=12, color='blue')
plt.ylabel("Exports of goods and services", fontsize=12, color='blue')
plt.xlabel("Country", fontsize=12, color='blue')
plt.grid(True, alpha=0.4)
plt.show()
```
