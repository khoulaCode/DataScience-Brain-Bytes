4. Compare the sharpened image to the original.

5. Save the processed image.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import PIL as Image
```

### **1. Load the provided X-ray image**

```python
X_ray = cv2.imread(r"C:\Users\bbuser\Desktop\DataScience-Brain-Bytes\Team_members\from_Hajer\1896_x-rays_of_Physikalischer_Verein,_hand_of_man.tif")
```

```python
#3 → number of channels (usually color channels: B, G, R in OpenCV)
X_ray.shape
```

```python
RGB_X_ray=cv2.cvtColor(X_ray, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(RGB_X_ray)
plt.title('X_ray image', fontsize=16)
plt.axis('off')
plt.show()
```

```python
X_ray.shape
```

### **2. Convert to grayscale (if not already)**

```python
Gray_X_ray = cv2.cvtColor(RGB_X_ray, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(Gray_X_ray, cmap='gray')
plt.title('Gray scale image' , fontsize=16)
plt.axis('off')
plt.show()
```

### **3. Apply a sharpening kernel (e.g., [[0,-1,0], [-1,5,-1], [0,-1,0]])**

```python
kernel=np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
sharped_X_ray=cv2.filter2D(Gray_X_ray,-1,kernel)
```

```python
def plot_image (sharped_X_ray, Gray_X_ray, title1 ,title2):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(sharped_X_ray, cmap='gray')
    plt.title(title1)
    plt.axis("off")
    
    plt.subplot(1,2,2)
    plt.imshow(Gray_X_ray , cmap='gray')
    plt.title(title2)
    plt.axis("off")
    
    plt.show()
```

```python
plot_image(sharped_X_ray, Gray_X_ray, "Sharped Image","Original")
```

### **5. Save the processed image**

```python
cv2.imwrite("sharpened_xray.png", sharped_X_ray)
```
