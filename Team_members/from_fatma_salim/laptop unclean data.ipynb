## convert frpm object to int in Ram :

```python
df['Ram'] = df['Ram'].astype(str).str.replace('GB', '', regex=False).astype(int)
df
```

## convert weight column from str to float :

```python
df['Weight'] = df['Weight'].astype(str).str.replace('kg', '', regex=False).str.strip()

df['Weight'] = df['Weight'].replace('?', np.nan)
df['Weight'] = df['Weight'].astype(float)
df
```

```python
df['Inches'].unique()
```

```python
df = df[df['Inches'].astype(str).str.match(r'^\d+(\.\d+)?$')]
df['Inches'] = df['Inches'].astype(float)
df
```

```python
df['MemorySize'] = df['Memory'].astype(str).str.extract(r'(\d+)').astype(float)
df
```

```python
df.drop_duplicates(inplace=True) #delete dublicate column 

print(" Unique values per column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")

print(f"\n Data shape after cleaning: {df.shape}")
```

```python
df
```

```python
text_columns = ['Memory', 'Cpu', 'ScreenResolution', 'Gpu', 'OpSys', 'Company', 'TypeName']

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

text_columns
```

```python
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.replace(r'\s+', ' ', regex=True)
    print(f"Cleaned values in column '{col}':")
    print(df[col].head())
    print('-' * 40)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['MemorySize'])
```

```python
mean_memory = df['MemorySize'].mean()
df.loc[:, 'MemorySize'] = df['MemorySize'].fillna(mean_memory)
mean_memory
```

```python
plt.boxplot(df['Price'].dropna())
plt.title('Boxplot of Price')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(df['Ram'], df['Price'], alpha=0.5)
plt.title('Scatter Plot of Ram vs Price')
plt.xlabel('Ram (GB)')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(df['Weight'], df['Price'], alpha=0.5, color='teal')
plt.title('Scatter Plot: Laptop Weight vs Price')
plt.xlabel('Weight (kg)')
plt.ylabel('Price')
plt.grid(alpha=0.7)
plt.show()
```

```python
plt.hist(df['Weight'].dropna(), bins=20, color='lightblue', edgecolor='black')
plt.title('Histogram of Laptop Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Count')
plt.show()
```
