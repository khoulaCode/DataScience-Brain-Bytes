```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the dataset

```python


df = pd.read_csv('household_power_consumption.txt', sep=';')
df
```

# pd.to_datetime() conversion

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'],dayfirst=True)
df
```

# Setting datetime index

```python
df.set_index('Datetime', inplace=True)
```

# Convert power usage column to numeric (handle missing values)

```python
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'])
df
```

# Drop rows with missing values

```python
df.dropna(subset=['Global_active_power'], inplace=True)
```

# Resample the data

```python
# Daily average
daily = df['Global_active_power'].resample('D').mean()
daily
```

```python
# Weekly average
weekly = df['Global_active_power'].resample('W').mean()
weekly
```

```python
# monthly average
monthly = df['Global_active_power'].resample('ME').mean()
monthly
```

# slicing df.loc()

```python
january_2007 = df.loc['2007-01']
january_2007
```

```python
specific_day = df.loc['2007-02-01']
specific_day
```

# plotting 

```python
daily.plot(title='Daily Average Power Consumption', figsize=(10,5), color="red", alpha=0.6)
plt.ylabel('Global Active Power (kilowatts)')
plt.grid(True, linestyle=":", alpha=0.6)
plt.show()
```

## ExpectFrom the daily average power usage plot:
I observe a seasonal pattern with higher consumption in the winter months. There is a gradual decline in average power consumption after 2009, possibly due to energy efficiency measures. A few outlier spikes in mid-2008 may suggest unusual events or high energy demands."

```python
plt.figure(figsize=(10,5))
plt.plot(weekly, label='Weekly Avg', color='orange')
plt.title('Weekly Average Power Consumption')
plt.ylabel('Global Active Power (kilowatts)')
plt.legend()
plt.grid(True)
plt.show()
```

## Expact from Weekly plots: 

*A very high week might indicate an event with high electricity use.

*A very low week could mean equipment was off, or a blackout.



```python
monthly.plot(title='Monthly Average Power Consumption', figsize=(10,5), color="gray")
plt.ylabel('Global Active Power (kilowatts)')
plt.grid(True, alpha=0.5)
plt.show()
```

## expact from monthly plot:

*People using energy-efficient devices

*Behavioral changes (less usage)

*External factors like policy or pricing
