# plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x= np.linspace(0,10,100)
y= np.sin(x)
y2=np.cos(x)
x
y
y2
```

```python
plt.figure(figsize=(20,10)) #hight first then length 
plt.plot(x, y , label="sin(x)" ,color="red", linestyle="--") #doted another linestyle
plt.plot(x, y2 , label="cos(x)" ,color="blue", linestyle="--")
plt.title("sin and cosine waves", fontsize=20)
plt.legend () # need to write for print the labels of label="sin(x)" abd cos(x)
plt.xlabel("x-axis" , fontsize= 20) #labela of fig
plt.ylabel("y-axis", fontsize=20)
plt.grid(True) #grid of fig inside the draw
plt.show () #to show the plot (line plot for time serise )
```

```python
np.random.seed(42)
x_scatter=np.random.rand(50) *10 # 0 to 1 but if * so give to 10
y_scatter=np.random.rand(50) *10 + x_scatter *0.5 + np.random.randn(50) # randn normal distrubation

plt.figure(figsize=(15,8))
plt.scatter(x_scatter, y_scatter , color="red", alpha=0.6 , edgecolors="black") #edgecolor mean setting lines of dots
plt.title("scatter plot of X vs Y")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True, linestyle=":", alpha=0.6) #alpha mean transparency of grid
plt.show()
```

```python
categories= ['a','b','c','d','e'] #product values
values=[23,45,65,18,39] #numircal values
plt.figure(figsize=(15,8))
plt.bar(categories, values, color=["red" ,"blue", "yellow", "black", "skyblue"]) #bar chart to compare category values
plt.title("Bar Plot of Category Values")
plt.xlabel("category")
plt.ylabel("values")
plt.grid(True , alpha= 0.4)
plt.show()
```

# histogram 

```python
data_hist= np.random.randn(1000)*15 +50 #normal dist number
data_hist_2= np.random.randn(1000)*15 + 30
plt.figure(figsize=(15,8))
plt.hist(data_hist , bins=10, color="teal", edgecolor="black", alpha=0.4)
plt.hist(data_hist_2 , bins=10, color="red", edgecolor="black", alpha=0.4)
plt.title("histgoram of data distrubuation ")
plt.xlabel("value")
plt.ylabel ("frequancey")
plt.savefig("figure.png") # to save it as image 
plt.show()
```

```python
data_to_plot= [data_hist, data_hist_2]
labels=["group1" , "group2"]
plt.figure(figsize=(6,10))
plt.boxplot(data_to_plot, 
            tick_labels=labels, 
            patch_artist=True, 
            boxprops=dict(facecolor="lightblue"))

plt.title("box plot of data groups")
plt.xlabel("group")
plt.ylabel("value")
plt.grid(axis="y" , linestyle="--" , alpha=0.5)
plt.show()
```
