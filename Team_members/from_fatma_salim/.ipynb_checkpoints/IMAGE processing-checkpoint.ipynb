## Solving task:  Understanding Digital Images with Python

##  1. Load and View a Grayscale Image

```python
import cv2
import matplotlib.pyplot as plt

# Load image in grayscale
img = cv2.imread("image3.jpg", cv2.IMREAD_GRAYSCALE)

# Display image
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()

# Check properties
print(f"Shape: {img.shape}")
print(f"Data type: {img.dtype}")
print(f"Min/Max pixel value: {img.min()} / {img.max()}")
```

## 2. Brighten the Image

```python
bright_img = img + 50
plt.imshow(bright_img, cmap='gray')
plt.title("Brightened (+50)")
plt.axis('off')
plt.show()
```

## 3. Add a Higher Value

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert to int before adding
too_bright = img.astype(np.int16) + 300  # Prevent overflow
too_bright = np.clip(too_bright, 0, 255).astype(np.uint8)  # Clip and convert back

# Show result
plt.imshow(too_bright, cmap='gray')
plt.title("Brightened Correctly (+300)")
plt.axis('off')
plt.show()

```

## 4. Subtract to Darken

```python
dark_img = img - 100
plt.imshow(dark_img, cmap='gray')
plt.title("Darkened (-100)")
plt.axis('off')
plt.show()

# Fix underflow
dark_img_fixed = np.clip(img - 100, 0, 255).astype(np.uint8)
plt.imshow(dark_img_fixed, cmap='gray')
plt.title("Darkened with Clipping")
plt.axis('off')
plt.show()
```

## 5. Transpose the Image

```python
transposed = img.T
plt.imshow(transposed, cmap='gray')
plt.title("Transposed Image")
plt.axis('off')
plt.show()
```

## 7. Slice (Crop) a Section of the Image

```python
flip_lr = np.fliplr(img)
flip_ud = np.flipud(img)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(flip_lr, cmap='gray')
plt.title("Flipped Left-Right")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(flip_ud, cmap='gray')
plt.title("Flipped Up-Down")
plt.axis('off')
plt.show()
```

## 8. Create a Negative Image

```python
negative = 255 - img
plt.imshow(negative, cmap='gray')
plt.title("Negative Image")
plt.axis('off')
plt.show()
```

## 9. Rotate the Image

```python
rotated = np.rot90(img, k=1)  # 90 degrees counter-clockwise
plt.imshow(rotated, cmap='gray')
plt.title("Rotated 90°")
plt.axis('off')
plt.show()
```

## 10. Combine Multiple Transformations

```python
# Get height and width
h, w = img.shape

# Step 1: Crop center 100x100
crop = img[h//2-50:h//2+50, w//2-50:w//2+50]

# Step 2: Flip horizontally
flipped = np.fliplr(crop)

# Step 3: Brighten
brightened = np.clip(flipped.astype(np.int16) + 80, 0, 255).astype(np.uint8)

# Show final result
plt.imshow(brightened, cmap='gray')
plt.title("Cropped → Flipped → Brightened")
plt.axis('off')
plt.show()
```

## Bouns challange 

```python
downsampled = img[::2, ::2]
collage = np.hstack([
    img[0:200, 0:200],
    255 - img[0:200, 0:200],
    np.fliplr(img[0:200, 0:200]),
    downsampled[0:200, 0:200]
])

plt.imshow(collage, cmap='gray')
plt.title("Collage: Original | Negative | Flip | Downsampled")
plt.axis('off')
plt.show()
```
