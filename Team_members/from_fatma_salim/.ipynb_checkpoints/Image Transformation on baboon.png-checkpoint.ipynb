```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Load the RGB Image

```python
img = cv2.imread('baboon.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def plot_image(images, titles, cmap=None):
    plt.figure(figsize=(15, 5))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        if cmap:
            plt.imshow(img, cmap=cmap)
        else:
            plt.imshow(img)
        plt.title(title)
        plt.axis('off')
    plt.show()
```

## Display Images + RGB Histograms

```python
def plot_rgb_histograms(image_rgb):
    colors = ('r', 'g', 'b')
    plt.figure(figsize=(10, 4))
    for i, color in enumerate(colors):
        hist = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
        plt.plot(hist, color=color)
        plt.xlim([0, 256])
    plt.title("RGB Channel Histograms")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.show()

plot_image([img_rgb], ["Original RGB Image"])
plot_rgb_histograms(img_rgb)
```

## Intensity Transformations

## Brightness Adjustment

```python
bright_img = cv2.convertScaleAbs(img_rgb, alpha=1, beta=50)  # beta controls brightness
```

## Contrast Adjustment

```python
contrast_img = cv2.convertScaleAbs(img_rgb, alpha=1.5, beta=0)  # alpha controls contrast
```

```python
plot_image([img_rgb, bright_img, contrast_img], ["Original", "Brightened", "Increased Contrast"])
```

 ## Histogram Equalization (Per Channel)

```python
def equalize_rgb(image_rgb):
    channels = cv2.split(image_rgb)
    eq_channels = [cv2.equalizeHist(cv2.cvtColor(c[..., None], cv2.COLOR_RGB2GRAY)) if c.ndim == 2 else cv2.equalizeHist(c) for c in channels]
    return cv2.merge(eq_channels)

img_ycrcb = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2YCrCb)
img_ycrcb[:, :, 0] = cv2.equalizeHist(img_ycrcb[:, :, 0])
equalized_img = cv2.cvtColor(img_ycrcb, cv2.COLOR_YCrCb2RGB)

plot_image([img_rgb, equalized_img], ["Original", "Histogram Equalized"])
plot_rgb_histograms(equalized_img)
```

## Convert to Grayscale + Histogram

```python
gray_img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(6, 4))
plt.hist(gray_img.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

plot_image([gray_img], ["Grayscale Image"], cmap='gray')
```

## Otsu’s Thresholding

```python
_, otsu_thresh = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plot_image([gray_img, otsu_thresh], ["Grayscale", "Otsu Threshold"], cmap='gray')
```
