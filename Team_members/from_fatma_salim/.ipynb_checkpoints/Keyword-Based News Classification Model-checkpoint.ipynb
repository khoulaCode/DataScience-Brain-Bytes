```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from wordcloud import WordCloud
from collections import Counter
import re
```

## Step 1: Load and Prepare Dataset

```python
# Load dataset from Kaggle input path
df = pd.read_json("News_Category_Dataset_v3.json", lines=True)

# Keep only the required categories
categories = ["POLITICS", "TRAVEL", "SPORTS", "HOME & LIVING"]
df = df[df['category'].isin(categories)]

# Keep only short_description and label
df = df[['short_description', 'category']]

# Balance dataset: sample 1000 per category (fixed to avoid warning)
df_balanced = df.groupby('category', group_keys=False).sample(n=1000, random_state=42)

# Stratified 80/20 split
train_df, test_df = train_test_split(
    df_balanced,
    test_size=0.2,
    stratify=df_balanced['category'],
    random_state=42
)

print("Dataset shape:", df_balanced.shape)
print("Train size:", train_df.shape, "Test size:", test_df.shape)
train_df.head()
```

## Step 2: Text Preprocessing

```python
def preprocess_text(text):
    # Lowercase & keep only alphabetic words
    return re.findall(r'\b[a-z]+\b', text.lower())
```

## Step 3: Word Frequencies and  Word Clouds

```python
category_word_freq = {}

for cat in categories:
    texts = train_df[train_df['category'] == cat]['short_description']
    all_words = []
    for t in texts:
        all_words.extend(preprocess_text(t))
    
    # Count word frequencies
    word_counts = Counter(all_words)
    category_word_freq[cat] = word_counts
    
    # Generate word cloud
    wc = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(word_counts)
    plt.figure(figsize=(8,4))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"WordCloud - {cat}", fontsize=16)
    plt.show()
```

## Step 4: Get Top Frequent Words per Category

```python
def get_top_words(freq_dict, N):
    return set([w for w, _ in freq_dict.most_common(N)])

# Example: top 10 words for each category
for cat in categories:
    top10 = get_top_words(category_word_freq[cat], 10)
    print(cat, ":", top10)
```

## Step 5: Keyword-Based Classifier

```python
def classify_text(text, top_words_dict):
    words = set(preprocess_text(text))
    overlaps = {cat: len(words & top_words) for cat, top_words in top_words_dict.items()}
    
    # Find category with max overlap
    max_overlap = max(overlaps.values())
    best_cats = [cat for cat, count in overlaps.items() if count == max_overlap]
    
    if len(best_cats) == 1:
        return best_cats[0]
    else:
        return "UNKNOWN" 
```

## Step 6: Evaluate for Different N (10, 20, 30, 50)

```python
def evaluate_classifier(N):
    # Get top N words per category
    top_words_dict = {cat: get_top_words(freqs, N) for cat, freqs in category_word_freq.items()}
    
    predictions = test_df['short_description'].apply(lambda x: classify_text(x, top_words_dict))
    accuracy = (predictions == test_df['category']).mean()
    return accuracy

results = {}
for N in [10, 20, 30, 50]:
    acc = evaluate_classifier(N)
    results[N] = acc
    print(f"Top {N} words -> Accuracy: {acc:.4f}")
```

```python
plt.plot(list(results.keys()), list(results.values()), marker='o')
plt.xlabel("Number of frequent words per category", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.title("Accuracy vs Number of Frequent Words", fontsize=14)
plt.grid(True)
plt.show()
```
