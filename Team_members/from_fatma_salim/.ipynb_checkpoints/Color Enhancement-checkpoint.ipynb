## Objective:
Automatically enhance the colors and contrast of landscape photos to make them more vivid and visually striking. This involves improving brightness and contrast through histogram equalization and boosting the color intensity by increasing saturation, thereby bringing out the true beauty of the scenery without distorting the natural hues.



## Load the input image in BGR or RGB format.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("zeeland.jpg")
image.shape
```

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

## Contrast Enhancement:

-Convert the image from BGR (or RGB) to HSV color space.

-Extract the V (Value) channel which represents brightness.

-Apply histogram equalization to the V channel to improve overall brightness and contrast.

-Replace the original V channel in the HSV image with the equalized version.


```python
# Extract the V (brightness) channel
v_channel = hsv_image[:, :, 2]

# Apply histogram equalization
equalized_v = cv2.equalizeHist(v_channel)

alpha = 0.7  # 0.0 = no change, 1.0 = full equalization
v_channel = cv2.addWeighted(equalized_v, alpha, v_channel, 1 - alpha, 0)


hsv_image[:, :, 2] = v_channel


contrast_enhanced_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
```

## Saturation Enhancement:

Convert the original image to HSV (if not already).

Extract the S (Saturation) channel.

Multiply the saturation channel by a factor to increase color intensity.

Use clipping to make sure values remain within the valid range [0, 255] to prevent artifacts.

Replace the original saturation channel with the enhanced one.

Convert the HSV image back to BGR (or RGB).

```python
# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Change the multiplier here:
saturation_boost = 1.2  # 20% boost (milder)
# saturation_boost = 2.0  # 100% boost (very strong)

# Apply boost and clip
hsv_image[:, :, 1] = np.clip(hsv_image[:, :, 1] * saturation_boost, 0, 255)

# Convert back to BGR
saturation_enhanced_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
```

## Display the enhanced images to compare with the original

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(contrast_enhanced_image, cv2.COLOR_BGR2RGB))
plt.title('Contrast Enhanced')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(saturation_enhanced_image, cv2.COLOR_BGR2RGB))
plt.title('Saturation Enhanced')
plt.axis('off')

plt.tight_layout()
plt.show()

```
