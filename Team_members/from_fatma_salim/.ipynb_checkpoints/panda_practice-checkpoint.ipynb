# Groupby
~ this will group object that holds all rows grouped by 'Country'.

```python
grouped = df.groupby(" Country ")
df
```

# GroupBy: Average GDP per country

```python
avg_gdp = df.groupby(" Country ")[" Gross Domestic Product (GDP) "].mean()
avg_gdp
```

# GroupBy + max: Highest GDP per country

```python
max_gdp = df.groupby(' Country ')[" Gross Domestic Product (GDP) "].max()
max_gdp
```

 # GroupBy + min: Lowest GNI per country

```python
min_gni = df.groupby(' Country ')[" Gross National Income(GNI) in USD "].min()
min_gni
```

# GroupBy + agg: Multiple aggregations at once

```python
agg_result = df.groupby(' Country ').agg({ " Gross Domestic Product (GDP) ": ['min', 'max', 'mean'],
    " Gross National Income(GNI) in USD ": ['min', 'max', 'mean'],
    " Population ": ['max']}).reset_index()

agg_result
```

# merge convert to nummirc value 

```python
df[' Gross Domestic Product (GDP) '] = pd.to_numeric(df[' Gross Domestic Product (GDP) '])
df
```

# use groupby then merge: GDP + GNI 

```python
avg_gdp = df.groupby(" Country ")[[" Gross Domestic Product (GDP) "]].mean()
avg_gni = df.groupby(" Country ")[[" Gross National Income(GNI) in USD "]].mean()

merged_gdp_gni = pd.merge(avg_gdp, avg_gni, on=" Country ")

merged_gdp_gni
```

# This merge :
connects each country to its region and adds population data.

# Benefit:
It is analyze economic indicators by region (compare Asia vs South America).

```python
left = pd.DataFrame({
    " Country_Name ": [" India ", " China ", " Brazil "],
    " Region ": [" Asia ", " Asia ", " South America "]})

right = df.groupby(" Country ")[[" Population "]].mean().reset_index()

merged_diff_names = pd.merge(left, right, left_on=" Country_Name ", right_on=" Country ")

merged_diff_names
```

# Outer Join — keep all from both

```python
merged_outer = pd.merge(avg_gdp, avg_gni, on=" Country ", how="outer")
merged_outer
```

# merge from left 

```python
merged_left = pd.merge(left, right, left_on=" Country_Name ", right_on=" Country ", how="left")
merged_left
```

# Combine India and China rows
Result: All rows from India + China stacked together.

```python
india = df[df[' Country '] == ' India ']
china = df[df[' Country '] == ' China ']

combined = pd.concat([india, china], axis=0)
combined
```

# Combine columns (same rows)

Result: Year and GDP side by side.

```python
a = df[df[' Country '] == ' India '][[' Year ']]
b = df[df[' Country '] == ' India '][[' Gross Domestic Product (GDP) ']]

side_by_side = pd.concat([a, b], axis=1)

side_by_side
```

# Pivot Table
reshapes the data:
Rows → become categories (like Year)
Columns → become grouped items (like Country)
Values → become numbers (like GDP)

```python
pivot_gdp = df.pivot(index=' Year ', columns=' Country ', values=' Gross Domestic Product (GDP) ')
pivot_gdp
```

# pivot table : population 
year (rows)
country (columns)
population (values)

```python
pivot_pop = df.pivot(index=' Year ', columns=' Country ', values=' Population ')
pivot_pop
```
