# 1. Top 5 Countries by Average GDP

```python
topAvg = df.groupby("Country")["Gross Domestic Product (GDP)"].mean().sort_values(ascending=False).head(5)
print("The Top 5 Countries " , topAvg)
```

# 2. Countries with Household Consumption Expenditure (most recent year)

```python
year = df['Year'].max()
consumption = df[df['Year'] == year][['Country', 'Household consumption expenditure (including Non-profit institutions serving households)']]
consumptionSorted = consumption.sort_values(by='Household consumption expenditure (including Non-profit institutions serving households)', ascending=False)
print(consumptionSorted)
```

```python
df.columns = df.columns.str.strip() 
meanG = df.groupby('Country')["Per capita GNI"].mean()
top10 = meanG.sort_values(ascending=False).head(10)
print(top10)
```

```python
maxImports = df.groupby("Country")["Imports of goods and services"].max()
imports10 = maxImports.sort_values(ascending=False).head(10)
print(imports10)
```

```python
totalExports = df.groupby("Country")["Exports of goods and services"].sum()
exports10 = totalExports.sort_values(ascending=False).head(10)
print(exports10)
```

```python
plt.figure(figsize=(10, 6))
topAvg.plot(kind='bar', color='powderblue')
plt.title('Top 5 Countries by Average GDP')
plt.xlabel('Country')
plt.ylabel('Average GDP')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
Consumption = consumptionSorted.head(12)
plt.figure(figsize=(12, 6))
plt.bar(Consumption['Country'], Consumption['Household consumption expenditure (including Non-profit institutions serving households)'], color='skyblue')
plt.title(f'Household Consumption Expenditure (Year: {year})')
plt.xlabel('Country')
plt.ylabel('Consumption Expenditure')
plt.xticks(rotation=45, ha='right')  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top10.sort_values().plot(kind='barh', color='lightgreen', edgecolor='black')
plt.title('Top 10 Countries by Mean Per Capita GNI')
plt.xlabel('Per Capita GNI (USD)')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
imports10.sort_values().plot(kind='barh', color='#2ca02c', edgecolor='black')
plt.title('Top 10 Importers (Goods & Services)', fontsize=14)
plt.xlabel('Imports (USD)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
exports10.plot(kind='bar', color='#3498db', edgecolor='navy', alpha=0.8)
plt.title('Top 10 Exporters of Goods and Services (Total)', fontsize=14, pad=20)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Exports (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
```
