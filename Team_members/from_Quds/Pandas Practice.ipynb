###  1. Basic insights

##### Top 5 countries by average GDP.

```python
# Group the data by 'Country' and calculate mean GDP for each
Top5_GDP = (df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending=False).head()) 
# Compute the average GDP for each country then  Sort in descending order ,# Get the top 5 countries
Top5_GDP
```

```python
# Plot a bar chart of top 5 countries by average GDP
Top5_GDP.plot(kind='bar', figsize=(6,6), color=['Skyblue','Lightcoral', 'lightgreen','lightpink','purple']) #Bar chart is used for comparing categories.
plt.title("Top 5 Countries by Average GDP")
plt.ylabel("Average GDP")
plt.xlabel("Country")
plt.xticks(rotation=45)  # Rotate country names for better readability
plt.show()
```

### 2. Countries with Household consumption expenditure in the most recent year

```python
print(df.columns.tolist())  #to see the columns
```

```python
# 2. Countries with Household consumption expenditure in the most recent year
most_recent_year = df[' Year '].max()  #find the year(recent year
hce_recent = df[df[' Year '] == most_recent_year][[' Country ', ' Household consumption expenditure (including Non-profit institutions serving households) ']]
#print(f"Household consumption expenditure in {most_recent_year}:\n", hce_recent)  #hce_recent -->Household consumption expenditure
hce_recent
```

```python
# Clean up column name
hce_recent.columns = ['Country', 'Household Consumption Expenditure']

# Sort by expenditure (optional)
hce_recent = hce_recent.sort_values('Household Consumption Expenditure', ascending=False).head()

# Plot
plt.figure(figsize=(6,6))
plt.bar(hce_recent['Country'], hce_recent['Household Consumption Expenditure'], color=['Skyblue','Lightcoral', 'lightgreen','lightpink','purple'])
plt.title(f'Household Consumption Expenditure by Country in {most_recent_year}')
plt.xlabel('Country')
plt.ylabel('Household Consumption Expenditure')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Group by Country and calculate:
Mean GDP per capita.
Max import of goods values
Total exports over all years.

```python
# 3. Group by Country and calculate:  for each i have specified the aggeragation and then rename the columns
grouped = df.groupby(' Country ').agg({ ' Gross Domestic Product (GDP) ': 'mean',' Imports of goods and services ': 'max',' Exports of goods and services ': 'sum'}).rename(
    columns={'GDP per capita': 'Mean GDP per capita',
    'Import of goods values': 'Max Import of goods',
    'Exports': 'Total Exports'}).head()  # only to get the first 5
grouped
```

```python
grouped.plot(kind='bar', stacked=True, figsize=(10, 6),color=['purple','DarkBlue','lightpink'])
plt.title('Mean GDP, Max Imports, and Total Exports by Country')
plt.ylabel('Value')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
