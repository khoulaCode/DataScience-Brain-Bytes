###  <span style="color:darkmagenta">1. Aggregation with groupby()</span>

#### <span style="color:deeppink">1.1 Total sales by store</span>

###### groupby('Store') splits the data for each Store.

###### sum() calculates total sales per store.

###### reset_index() flattens the result back into a DataFrame.

```python
Sales_By_Store = train_data_df.groupby('Store')['Weekly_Sales'].sum().reset_index()
Sales_By_Store.rename(columns={'Weekly_Sales': 'Total_Sales'}, inplace=True)  # to rename the columns in the train tale( from weekly sales to total sales)
# rename() creates a new DataFrame where the columns (or index) are renamed.
# If you don’t tell it to inplace, the original DataFrame isn’t changed.. instead, it returns a copy with the changes.
# Why??? Because inplace=True tells pandas:"Modify this DataFrame directly, don’t create a copy."
print(Sales_By_Store.sort_values('Total_Sales', ascending=False))
```

### <span style="color:deeppink">1.2 Average Sales by Department</span>

```python
Average_Sales_By_Department = train_data_df.groupby('Dept')['Weekly_Sales'].mean().reset_index()  #find the average(mean) 
Average_Sales_By_Department.rename(columns={'Weekly_Sales': 'Average_Sales(weekly)'}, inplace=True)  # to rename the columns in the train tale( from weekly sales to Average sales)
Average_Sales_By_Department
```

### <span style="color:deeppink">1.3 Sales Trend Over Time</span>

```python
train_data_df['Date'] = pd.to_datetime(train_data_df['Date'])
weekly_trend = train_data_df.groupby('Date')['Weekly_Sales'].sum().reset_index()
print("Sales Trend:\n")
weekly_trend
```

###  <span style="color:darkmagenta">2. Merging different datasets using: merge()</span>

```python
#merge(): Combine two Series or DataFrame objects with SQL-style joining
```

#### <span style="color:deeppink">2.1 Merge sales with features</span>

```python
train_features = pd.merge(train_data_df, features_data_df,  on=['Store','Date'], how='left') 
print("Merge sales with features:\n")
train_features
```

```python
# there is an error because for the date col is having  the date as an object  so i have to change the date for the both tables   as datetime
#Convert 'Date' in both DataFrames to datetime
```

```python
train_data_df['Date'] = pd.to_datetime(train_data_df['Date'])
features_data_df['Date'] = pd.to_datetime(features_data_df['Date'])
```

#### <span style="color:deeppink">2.2 Merge with store information</span>

```python
# to  have one single dataset combining sales,features,and store datas.
stores_info = pd.merge(stores_data_df, features_data_df,  on='Store', how='left')
print("Merge with store information:\n")
stores_info
```

### <span style="color:darkmagenta">3. Concatenating data</span>

```python
#concat(): Merge multiple Series or DataFrame objects along a shared index or column
#Why use concat:Stack rows vertically (like appending one DataFrame below another).
```

```python
# Split into two time periods
First_Date_half = train_data_df[train_data_df['Date'] <= '2012-05-15']
second_Date_half = train_data_df[train_data_df['Date'] > '2012-05-15']

# Concatenate back
combined_timePeroid_df = pd.concat([First_Date_half, second_Date_half])
combined_timePeroid_df
```
