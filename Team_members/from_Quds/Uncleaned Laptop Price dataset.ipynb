## Data Cleaning

```python
df_process = df.drop('Unnamed: 0',axis = 1)   # it deletes the column that have the null values
df_process
```

```python
df_process.isnull().sum()
```

```python
df_process.head(50)
```

```python
df_process.dropna(how = 'all', inplace = True)  # to drop the rows that have nan 
df_process
```

```python
df_process.isnull().sum()  # to check the null values after drop
```

```python
df_process.tail(50)
```

```python
df_process = df_process.reset_index(drop=True)  #-->reset the index to make it sequential..using drop means it will remove the old index instead of adding it as a new column.
df_process
```

```python
# Replace '?' with NaN  --> if i did not replace it i will not be able to convert it
df_process['Inches'] = df_process['Inches'].replace('?', np.nan)
```

```python
# Convert Inches to float
df_process['Inches'] = df_process['Inches'].astype(float)
```

```python
df_process.dtypes
```

```python
df_process['Ram'] = df_process['Ram'].str.replace('GB', '').astype(int) # Clean Ram and convert to int
```

```python

df_process['Weight'] = df_process['Weight'].replace('?', np.nan)
```

```python
df_process['Weight'] = df_process['Weight'].str.replace('kg', '').astype(float )# Clean Weight and convert to float
```

```python
df_process.dtypes
```

```python
df_process[df_process['Weight'].isnull()]
```

```python
# there is a relationship between Weight and Inches in laptops...
# In general, bigger screens (Inches) mean heavier laptops (Weight) because larger screens require bigger frames, batteries, and components
```

```python
# Compute mean Inches for each Weight value
inches_by_weight = df_process.groupby('Weight')['Inches'].transform('mean')

```

```python
# Fill missing Inches with the mean Inches for the same Weight
df_process['Inches'] = df_process['Inches'].fillna(inches_by_weight)
```

```python
df_process[df_process['Inches'].isnull()]   #to check if there is null values
```

```python
# Compute mean Weight for each Inches value
weight_by_inches = df_process.groupby('Inches')['Weight'].transform('mean')
```

```python
# Fill missing Weight with the mean Weight for the same Inches
df_process['Weight'] = df_process['Weight'].fillna(weight_by_inches)
```

```python
plt.figure(figsize=(6,5))
plt.hist(df_process['Inches'], bins=15, color='lightblue', edgecolor='deeppink')
plt.title('Distribution of Screen Sizes')
plt.xlabel('Screen Size (Inches)',fontsize = 10 , color = 'darkmagenta')
plt.ylabel('Number of Laptops',fontsize = 10 , color = 'darkmagenta')
plt.grid(True, alpha = 0.4 , linestyle =":")
plt.show()
```
