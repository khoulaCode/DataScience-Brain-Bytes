##### Load the provided X-ray image.
##### Convert to grayscale (if not already)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bgr = cv2.imread(r"C:\Users\bbuser\Downloads\x-rays.png")

if img_bgr is None:
    print("Image not found! Please check the path.")
else:
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    plt.imshow(img_gray, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')
    plt.show()
```

##### Apply a sharpening kernel (e.g., [[0,-1,0], [-1,5,-1], [0,-1,0]]).

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

sharpened = cv2.filter2D(img_gray, -1, kernel)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Grayscale")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image")
plt.axis('off')

plt.show()
```

##### Compare the sharpened image to the original.

##### I noticed that after applying the sharpening kernel, the image edges became more defined and clearer compared to the original grayscale image.  The sharpened image shows enhanced details, making features stand out better, while the original image appears softer with less pronounced edges. Also, I noticed that after applying the sharpening kernel, the image became brighter and lighter compared to the original grayscale image. The sharpened image appears clearer, while the original looks darker and less detailed.
