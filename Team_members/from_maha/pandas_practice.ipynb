# Mean GDP per capita.

```python
mean_gpd = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean()
mean_gpd
```

```python
mean_gpd.sort_values(ascending=False).head(5).plot(kind='bar', color='green')
plt.title('Top 10 Countries by Mean GDP')
plt.show()
```

# Max import of goods values

```python
max_import = df.groupby(' Country ')[' Imports of goods and services '].max()
max_import
```

```python
max_import_sorted = max_import.sort_values(ascending=False).head(5)
plt.figure(figsize=(12, 6))
max_import_sorted.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Maximum Imports of Goods and Services by Country')
plt.xlabel('Country')
plt.ylabel('Maximum Imports')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Total exports over all years

```python
total_export = df.groupby(' Country ')[' Exports of goods and services '].sum()
total_export
```

```python
total_export_sorted = total_export.sort_values(ascending=False).head(5)

plt.figure(figsize=(12, 6))
total_export_sorted.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Top 10 Total Exports by Country')
plt.xlabel('Country')
plt.ylabel('Total Exports')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle=':', alpha=0.7)
plt.show()
```
