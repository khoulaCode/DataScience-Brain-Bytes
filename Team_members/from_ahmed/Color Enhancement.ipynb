```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Typical_New_Zealand_scenic_walk_view.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Convert to HSV
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_img)
```

```python
v_eq = cv2.equalizeHist(v)

# Merge and convert back to RGB
hsv_eq = cv2.merge((h, s, v_eq))
contrast_enhanced_rgb = cv2.cvtColor(hsv_eq, cv2.COLOR_HSV2RGB)
```

```python
# Increase saturation
s_sat = np.clip(s.astype(np.float32) * 1.5, 0, 255).astype(np.uint8)
```

```python
# Merge new saturation with original H & V
hsv_sat = cv2.merge((h, s_sat, v))
saturation_enhanced_rgb = cv2.cvtColor(hsv_sat, cv2.COLOR_HSV2RGB)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(image_rgb)
axs[0].set_title('Original')
axs[0].axis('off')

axs[1].imshow(contrast_enhanced_rgb)
axs[1].set_title('Contrast Enhanced')
axs[1].axis('off')

axs[2].imshow(saturation_enhanced_rgb)
axs[2].set_title('Saturation Enhanced')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
