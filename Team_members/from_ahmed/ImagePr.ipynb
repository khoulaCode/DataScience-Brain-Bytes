```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = Image.open('Switch_TLOZ_TOTK_Screen_30.jpg')
print(f"PIL Image format:{img.format}, size: {img.size}, mode: {img.mode}")
```

```python
plt.imshow(img)
plt.title('Zelda')
plt.axis('off')
plt.show()
```
