```python
import pandas as pd
```

```python
df = pd.read_json('News_Category_Dataset_v3.json', lines=True)
```

### Data Preprocessing

```python
df.columns.str.lower().str.strip()
```

```python
dff = df.groupby('category').head(1000)[['headline', 'category']].reset_index(drop=True)
```

```python
dff
```

```python
dff['category'].value_counts()
```

## Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
resuelt = tfidf.fit_transform(dff['headline'])
```

```python
newdf = pd.DataFrame(resuelt.toarray(), columns=tfidf.get_feature_names_out())
newdf
```

```python
query = ['Apple releases new iPhone model']
query_vec = tfidf.transform(query)
# from ipywidgets import widgets
# from IPython.display import display

# text = widgets.Text(description="Search:")
# display(text)

# def on_submit(change):
#     query_vec = tfidf.transform([text.value])
#     print(query_vec)  # or run your similarity search

# text.on_submit(on_submit)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(query_vec, resuelt).flatten()
```

```python
import numpy as np
top_indices = np.argsort(cos_sim)[-10:][::-1]
for idx in top_indices:
    print(f"Headline: {dff['headline'].iloc[idx]}")
    print(f"Category: {dff['category'].iloc[idx]}")
    print(f"Score: {cos_sim[idx]:.3f}")
    print("-" * 50)
```
