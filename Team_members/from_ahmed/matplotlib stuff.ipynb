```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(x, y, label='sin(x)', color='red', linestyle=':')
plt.plot(x, y2, label='cos(x)', color='blue', linestyle=':')
plt.title("Sine and Cosine Maves", fontsize=20)  # Note: Typo in "Maves", should be "Waves"
plt.legend()
plt.xlabel("X-axis", fontsize=16)
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)
X_scatter =np.random.rand(50) *10
Y_scatter =np.random.rand(50) *10 + X_scatter *0.5 +np.random.randn(50)
plt.figure(figsize=(15, 8))
plt.scatter(X_scatter , Y_scatter, color="red",alpha = 0.9, edgecolors ="black" )
plt.title("scatter plot of X vs Y")
plt.xlabel("X values")
plt.ylabel("y values")
plt.grid(True, linestyle ="--",alpha =0.4)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 65, 18, 39]

plt.figure(figsize=(15, 8))
bars = plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom')
plt.title("Bar Plot of Category Values")
plt.xlabel("Category")
plt.ylabel("Values")
plt.grid(True, alpha = 0.4)
plt.show()
```

```python
data_hist = np.random.randn(1000)*15 +50
data_hist2 = np.random.randn(1000)*15 +30

plt.figure(figsize=(15, 8))
plt.grid(True, alpha = 0.9, linestyle =":")
plt.hist(data_hist, bins=30, color='black', edgecolor='black', alpha=0.8)
plt.hist(data_hist2, bins=30, color='red', edgecolor='black', alpha=0.7)
plt.title('Histogram of Data Distribution')
plt.xlabel('value')
plt.ylabel('frequency')
plt.savefig("hist.png", dpi=100)
plt.show()
```

```python
data_to_plot = [data_hist, data_hist2]
labels = ["Group 1", "Group 2"]

plt.figure(figsize = (18, 6))
plt.boxplot(data_to_plot,patch_artist = True, tick_labels = labels, boxprops=dict(facecolor="red"))
plt.title("Box plot of the data group")
plt.xlabel("Group")
plt.ylabel("Value")
plt.grid(True, alpha = 0.9, linestyle ="--", axis ="y")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2,  figsize = (12, 10))

#plot 1
axes[0, 0].plot(x, y, color = "red")
axes[0, 0].set_title("sine wave")
axes[0, 0].grid(True)

#plot 2
axes[0, 1].plot(x, y2, color = "blue")
axes[0, 1].set_title("cosine wave")
axes[0, 1].grid(True)

#plot 3
axes[1, 0].hist(data_hist, bins=30, color='black', edgecolor='black', alpha=0.8)
axes[1, 0].set_title("hitsogram")
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("y")
axes[1, 0].grid(True)

#plot 4
axes[1, 1].boxplot(data_to_plot,patch_artist = True, tick_labels = labels, boxprops=dict(facecolor="red"))
axes[1, 1].set_title("boxplot")
axes[1, 1].set_xlabel("x")
axes[1, 1].set_ylabel("y")
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
