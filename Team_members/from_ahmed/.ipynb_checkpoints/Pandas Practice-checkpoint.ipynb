```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Global Economy Indicators.csv')
```

### Top 5 countries by average GDP.

```python
AVG = df.groupby(" Country ")[" Gross Domestic Product (GDP) "].mean().sort_values(ascending=False)
GDP = AVG.head(5)
GDP
```

### Countries with the Household consumption expenditure in the most recent year.

```python
lastyear = df[df[' Year '] == df[' Year '].max()]
Countries = lastyear.groupby(' Country ')[' Household consumption expenditure (including Non-profit institutions serving households) '].sum().head(5)
Countries
```

### Mean GDP per capita.

```python
df[' GDP per capita '] = df[' Gross Domestic Product (GDP) '] / df[' Population ']
mean = df.groupby(' Country ')[' GDP per capita '].mean().head(10)
print('mean: ')
mean
```

### Max import of goods values

```python
imports = df.groupby(' Country ')[' Imports of goods and services '].max()
maximports = imports.sort_values(ascending=False)
maximports.head(5)
```

###  Total exports over all years.

```python
exports = df.groupby(' Country ')[' Exports of goods and services '].sum()
total_exports = exports.sort_values( ascending=False)
total_exports.head(5)
```

### Plots

```python
GDP.plot(kind="bar", title="Top 5 Countries by Average GDP")
plt.ylabel('Average GDP')
plt.show()

mean.sort_values(ascending=False).plot(kind='bar', title='Mean GDP per Capita by Country')
plt.ylabel('GDP per Capita')
plt.show()

maximports.sort_values(ascending=False).plot(kind='bar', title='Max Imports by Country')
plt.ylabel('Max Imports')
plt.show()

exports.sort_values(ascending=False).plot(kind='bar', title='Total Exports by Country')
plt.ylabel('Total Exports')
plt.show()
```
