```python
import numpy as np
import pandas as pd
```

```python
# Example DataFrame
data = {'A': [1, 2, np.nan, 4, 5, np.nan, 10 ],
        'B': [np.nan, 2, 3, 4, np.nan, np.nan, 10],
        'C': [10, 20, 30, 40, 50, np.nan, 10],
        'D': [10, 20, 30, 40, 50, 9, 10]}
df = pd.DataFrame(data)

print("DataFrame with NaNs:")
df
```

```python
#find out the percentage of null values
df.isnull().sum() / len(df) * 100
```

```python
#drop raws with null values
#df_dropped = df.dropna(axis = 0)
#axis is 0 by default 
df_dropped = df.dropna()
df_dropped
```

```python
#drop columns with null values
df_dropped = df.dropna(axis = 1)
df_dropped
```

```python
#drop the raws with all null values
df_dropped = df.dropna(how='all')
df_dropped
```

```python
#drop a raw/column with a specifec number of null values using thresh
df_dropped = df.dropna(thresh= 4, axis = 0)
df_dropped
```

```python
#replace null values with 0
df.fillna(0)
```

```python
#fill the null values with the median
df.fillna(df.median(numeric_only=True))
```

```python
#fill the null values with the mode
df.fillna(df.mode().iloc[0])
```

```python
#fill only the column you want to fill in this case its 'D'
df['D'] = df[['D']].fillna(df.mode().iloc[0])
df
```

```python
#fill the null values with value that comes before(bfill) or after(ffill) the null value
df.fillna(method='bfill')
```
