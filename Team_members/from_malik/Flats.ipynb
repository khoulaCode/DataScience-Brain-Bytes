```python
# multiplier map: 1 Lac = 100,000; 1 Crore = 10,000,000
unit_map = {'lac': 1e5, 'lakh': 1e5, 'cr': 1e7, 'crore': 1e7}

def convert_price_series(s: pd.Series) -> pd.Series:
    cleaned = (s.astype(str).str.lower().str.replace('[₹,]', '', regex=True).str.strip())
    # treat call for price / missing-looking as NaN
    cleaned = cleaned.replace(r'call for price', pd.NA, regex=True)

    parts = cleaned.str.extract(r'(?P<number>[\d\.]+)\s*(?P<unit>lac|lakh|cr|crore)?')
    number = pd.to_numeric(parts['number'], errors='coerce')
    multiplier = parts['unit'].map(unit_map).fillna(1)
    return number * multiplier

# apply it
df['price_num'] = convert_price_series(df['price'])
```

```python
unit_map = {'lac': 1e5, 'lakh': 1e5, 'cr': 1e7, 'crore': 1e7}
def convert_price_series(s: pd.Series) -> pd.Series:
    cleaned = (
        s.astype(str)
         .str.lower()
         .str.replace('[₹,]', '', regex=True)
         .str.strip()
    )
    cleaned = cleaned.replace(r'call for price', pd.NA, regex=True)
    parts = cleaned.str.extract(r'(?P<number>[\d\.]+)\s*(?P<unit>lac|lakh|cr|crore)?')
    number = pd.to_numeric(parts['number'], errors='coerce')
    multiplier = parts['unit'].map(unit_map).fillna(1)
    return number * multiplier

df['price_num'] = convert_price_series(df['price'])
```
