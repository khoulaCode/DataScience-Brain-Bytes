```python
gdp_avg = (df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending=False))

top5 = gdp_avg.head(5)
print(top5)
```

```python
latest_year = df[' Year '].max()
recent_df = df[df[' Year '] == latest_year]
top_household = recent_df[[' Country ', 
    ' Household consumption expenditure (including Non-profit institutions serving households) ']
].sort_values(by=' Household consumption expenditure (including Non-profit institutions serving households) ', 
              ascending=False).head(5)
print(f"Top 5 Countries by Household Consumption in {latest_year}:\n", top_household)
```

```python
grouped = df.groupby(' Country ').agg({
    ' Per capita GNI ': 'mean',
    ' Imports of goods and services ': 'max',
    ' Exports of goods and services ': 'sum'
}).rename(columns={
    ' Per capita GNI ': 'Mean GNI per Capita',
    ' Imports of goods and services ': 'Max Import of Goods',
    ' Exports of goods and services ': 'Total Exports'
})

print("\nGrouped Statistics by Country:\n", grouped.head())
```

### Plot 1: Top 5 Average GDP

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 12))

# Plot 1: Top 5 Average GDP
top5.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Top 5 Countries by Average GDP')
axes[0].set_ylabel('Average GDP (USD)')

# Plot 2: Household Consumption Expenditure
axes[1].bar(top_household[' Country '], 
            top_household[' Household consumption expenditure (including Non-profit institutions serving households) '],
            color='salmon')
axes[1].set_title(f'Top 5 Countries by Household Consumption in {latest_year}')
axes[1].set_ylabel('Household Consumption Expenditure (USD)')

plt.tight_layout()
plt.show()
```
