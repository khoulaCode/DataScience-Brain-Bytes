```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load grayscale image
img = Image.open("test_image.jpg").convert("L")   # put your image path here
img = np.array(img)

plt.imshow(img, cmap="gray"); plt.title("Original"); plt.show()
```

```python
# 2. Brighten
bright = np.clip(img + 50, 0, 255)
plt.imshow(bright, cmap="gray"); plt.title("Brightened +50"); plt.show()
```

```python
# 3. Brighten with high value (+300)
bright300 = np.clip(img.astype(np.int32) + 300, 0, 255).astype(np.uint8)
plt.imshow(bright300, cmap="gray")
plt.title("Brightened +300")
plt.show()
```

```python
# 4. Darken
dark = np.clip(img - 100, 0, 255)
plt.imshow(dark, cmap="gray"); plt.title("Darkened -100"); plt.show()
```

```python
# 5. Transpose
transposed = img.T
plt.imshow(transposed, cmap="gray"); plt.title("Transposed"); plt.show()
```

```python
# 6. Flips
fliplr = np.fliplr(img)
flipud = np.flipud(img)
plt.imshow(fliplr, cmap="gray"); plt.title("Flip Left-Right"); plt.show()
plt.imshow(flipud, cmap="gray"); plt.title("Flip Up-Down"); plt.show()
```

```python
# 7. Crop
h, w = img.shape
crop = img[h//2-50:h//2+50, w//2-50:w//2+50]
plt.imshow(crop, cmap="gray"); plt.title("Center Crop 100x100"); plt.show()
```

```python
# 8. Negative
negative = 255 - img
plt.imshow(negative, cmap="gray"); plt.title("Negative"); plt.show()
```

```python
# 9. Rotate
rot90 = np.rot90(img, k=1)
plt.imshow(rot90, cmap="gray"); plt.title("Rotate 90"); plt.show()
```

```python
# 10. Combination: crop  flip  brighten
combo = np.clip(np.fliplr(crop) + 40, 0, 255)
plt.imshow(combo, cmap="gray"); plt.title("Combo (CropFlipBrighten)"); plt.show()
```
