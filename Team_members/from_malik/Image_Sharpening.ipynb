```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("1896_x-rays_of_Physikalischer_Verein,_hand_of_man.tif.png")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernel = [[0, -1, 0],
          [-1, 5, -1],
          [0, -1, 0]]

kernel = np.array(kernel)
```

```python
sharpened = cv2.filter2D(gray, -1, kernel)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(gray, cmap="gray")
plt.axis("off")
```

```python
plt.subplot(1,2,2)
plt.title("Sharpened")
plt.imshow(sharpened, cmap="gray")
plt.axis("off")
plt.show()
```
