```python
import pandas as pd
df = pd.read_csv('data/laptopData.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)

print(df.columns)
```

```python
# 1. drop empty rows (choose how='all' or how='any')
df.dropna(how='all', inplace=True)

# 2. reset the index
df.reset_index(drop=True, inplace=True)

# check
print(df.head())
print(df.columns)
```

```python
df.head()
```

```python
df.tail()
```

```python
dup_count = df.duplicated().sum()
print(f"Duplicate rows: {dup_count}")
```

```python
# 1. Drop duplicates (keeps the first occurrence by default)
df.drop_duplicates(inplace=True)

# 2. Verify: there should be zero duplicate rows now
print(f"Remaining duplicates: {df.duplicated().sum()}")

# 3. (Optional) See how many rows remain
print(f"New row count: {len(df)}")
```

```python
# Create a boolean mask: True for rows that contain at least one "?"
mask = df.eq('?').any(axis=1)

# Drop those rows
df = df[~mask].copy()

# (Optional) Check it worked:
print((df == '?').any())   # should all be False
print(df.shape)            # row count should be reduced
```

```python
unique_ops = df['OpSys'].unique()
print(unique_ops)
```

```python
import matplotlib.pyplot as plt

os_counts = df['OpSys'].value_counts()

os_counts.plot(kind='bar')

plt.xlabel('Operating System')
plt.ylabel('Number of Laptops')
plt.title('Laptop Counts by OpSys')

plt.tight_layout()
plt.show()
```

```python
for col in df.select_dtypes('object'):
    df[col] = df[col].str.strip()
```

```python
for col in ['Company','TypeName','OpSys']:
    print(col, df[col].unique())
```

```python
# 6. Quick numeric summary
print(df[['Inches','Price']].describe())
```

```python
df['Resolution']   = df['ScreenResolution'].str.split().str[-1]
df[['X_res','Y_res']] = df['Resolution'].str.split('x', expand=True).astype(int)
df['Touchscreen']  = df['ScreenResolution'].str.contains('Touch', case=False)
```

```python
df['Ram_GB'] = df['Ram'].str.replace('GB','').astype(int)
```

```python
df['Weight_kg'] = df['Weight'].str.replace('kg','').astype(float)
```

```python
df['Inches'] = df['Inches'].astype(float)
```

```python
df['Price']  = df['Price'].astype(float)
```

```python
parts = df['Cpu'].str.rsplit(' ', n=1, expand=True)
df['Cpu_Model']    = parts[0]
df['Cpu_Speed_GHz']= parts[1].str.replace('GHz','').astype(float)
```

```python
parts = df['Memory'].str.split(' \+ ', n=1, expand=True)
parts.columns = ['Part1','Part2']

df[['Mem1_Size','Mem1_Type']] = parts['Part1'].str.split(' ', n=1, expand=True)
df[['Mem2_Size','Mem2_Type']] = parts['Part2'].str.split(' ', n=1, expand=True)

df[['Mem2_Size','Mem2_Type']] = df[['Mem2_Size','Mem2_Type']].fillna('Not Available')

print(df[['Memory','Mem1_Size','Mem1_Type','Mem2_Size','Mem2_Type']].head())
```
