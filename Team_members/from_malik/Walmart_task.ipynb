# Merge train + features on Store & Date

```python
df = pd.merge(df_train,df_features,on=['Store','Date'],how='left')
```

# Merge in store 

```python
df = pd.merge(df,df_stores,on='Store',how='left')
```

```python
print(df.info())
```

# Total sales by Store

```python
sales_by_store = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False)
print(sales_by_store.head())
```

# Total sales by dept

```python
sales_by_dept = df.groupby('Dept')['Weekly_Sales'].sum().sort_values(ascending=False)
print(sales_by_dept.head())
```

# Average sales by Store Type

```python
avg_sales_by_type = df.groupby('Type')['Weekly_Sales'].mean()
print(avg_sales_by_type)
```

# Count of each store Type

```python
print(df_stores['Type'].value_counts())
```

# Which weeks had unusually high sales

```python
high_sales = df[df['Weekly_Sales'] > 200_000]
print(high_sales[['Store','Dept','Date','Weekly_Sales']].head())
```
