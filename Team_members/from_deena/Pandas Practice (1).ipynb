```python
import matplotlib.pyplot as plt
```

### **Get basic insights**

#### Top 5 countries by average GDP.

```python
average_gdp = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean()
sorted_gdp = average_gdp.sort_values(ascending=False).head(5)
sorted_gdp
```

#### Countries with the Household consumption expenditure in the most recent year.

```python
latest_year =df[' Year '].max()
latest = df[df[' Year '] == latest_year]
latest
```

### **Group by**

#### Group by Country and calculate Mean GDP per capita.

```python
grouped_df=df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean()
grouped_df
```

```python
plt.figure(figsize =(15, 8))
grouped_df.plot.bar(' Country ', ' Gross Domestic Product (GDP) ', color='skyblue')
plt.title("Bar Plot of Country GDP", color='red')
plt.xlabel(" Country ", color='red')
plt.ylabel(" Gross Domestic Product (GDP) ", color='red')
plt.grid(True, alpha=0.15, color='gray')
plt.show()
```

#### Group by Country and calculate Max import of goods values

```python
df.groupby(' Country ')[' Imports of goods and services '].max()
```

```python
plt.figure(figsize =(15, 8))
grouped_df.plot.bar(' Country ', ' Imports of goods and services ', color='skyblue')
plt.title("Bar Plot of Country GDP", color='red')
plt.xlabel(" Country ", color='red')
plt.ylabel(" Gross Domestic Product (GDP) ", color='red')
plt.grid(True, alpha=0.15, color='gray')
plt.show()
```

#### Group by Country and calculate Total exports over all years.

```python
df.groupby(' Country ')[' Exports of goods and services '].sum()
```

```python
plt.figure(figsize =(15, 8))
grouped_df.plot.bar(' Country ', ' Exports of goods and services ', color='skyblue')
plt.title("Bar Plot of Country GDP", color='red')
plt.xlabel(" Country ", color='red')
plt.ylabel(" Gross Domestic Product (GDP) ", color='red')
plt.grid(True, alpha=0.15, color='gray')
plt.show()
```

```python
import numpy as np
```

```python
grouped_df=df.groupby(' Country ').agg({
    ' Gross Domestic Product (GDP) ':np.mean,
    ' Imports of goods and services ':np.max,
    ' Exports of goods and services ':np.sum
})
grouped_df
```
