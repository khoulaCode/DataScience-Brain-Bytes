```python
def memory_type(m):
    if 'SSD' in m:
        return 'SSD'
    elif 'HDD' in m:
        return 'HDD'
    elif 'Flash' in m:
        return 'Flash Storage'
    elif 'Hybrid' in m:
        return 'Hybrid'
    return 'Other'
laptop['Memory_Type'] = laptop['Memory'].apply(memory_type)
```
