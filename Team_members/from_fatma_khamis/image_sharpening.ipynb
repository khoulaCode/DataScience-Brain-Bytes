```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("xray.png")
img.shape
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray.shape
```

```python
kernal = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
sharped_image = cv2.filter2D(img_gray, -1, kernel=kernal)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title('sharpened image')
plt.imshow(sharped_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
cv2.imwrite("sharpened_image.png", sharped_image)
```
