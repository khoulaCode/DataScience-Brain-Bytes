```python
# Countries with the Household consumption expenditure in the most recent year.
economy['Year'].max()
```

```python
recent_year = economy[economy['Year']==economy['Year'].max()]
```

```python
df = recent_year[["Country","Household consumption expenditure (including Non-profit institutions serving households)"]]
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df
```

```python
grouped = economy.groupby('Country').agg(GDP_mean=('Gross Domestic Product (GDP)','mean'),max_import_goods=('Imports of goods and services', 'max'),total_exports=('Exports of goods and services', 'sum'))
grouped
```

```python
plt.figure(figsize=(10,6))
top_gdp.plot(kind='bar')
plt.title('Top 5 Countries by Average GDP (1970–2021)')
plt.ylabel('Average GDP')
plt.show()
```

```python
sorted_df = df.sort_values(by='Household consumption expenditure (including Non-profit institutions serving households)', ascending=False).head(5)
plt.figure(figsize=(10, 5))
plt.barh(sorted_df['Country'], sorted_df['Household consumption expenditure (including Non-profit institutions serving households)'], color='red')
plt.title('Household Consumption Expenditure by Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(grouped['GDP_mean'].nlargest(5).index[::-1], grouped['GDP_mean'].nlargest(5), color='limegreen')
plt.title('Top 10 Countries by Mean GDP')
plt.xlabel('Mean GDP per Capita (USD)')
plt.tight_layout()
plt.show()
```
