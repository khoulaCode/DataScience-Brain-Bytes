```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

1. Load and View a Grayscale Image

```python
img_cv = cv2.imread("Shuksan.jpg", cv2.IMREAD_GRAYSCALE)
print(f"PIL Image Format: {img_cv.dtype}, size: {img_cv.shape}, pixel range: {img_cv.min()} to {img_cv.max()}")
```

```python
plt.imshow(img_cv, cmap='gray')
plt.axis("off")
plt.title("grayscale image")
plt.show()
```

2. Brighten the Image

```python
bright_img = img_cv.astype(np.int32) + 50
bright_img = np.clip(bright_img, 0, 255).astype(np.uint8)
plt.imshow(bright_img, cmap='gray')
plt.axis("off")
plt.title("bright image +50")
plt.show()
```

3. Add a Higher Value

```python
higher_img = img_cv.astype(np.int32) + 300
higher_img = np.clip(higher_img, 0,255).astype(np.uint8)
plt.imshow(higher_img, cmap='gray')
plt.axis("off")
plt.title("bright image +300")
plt.show()
```

4. Subtract to Darken

```python
darker_img = img_cv.astype(np.int16) - 100
darker_img = np.clip(darker_img, 0, 255).astype(np.uint8)
plt.imshow(darker_img, cmap='gray')
plt.axis("off")
plt.title("dark image -100")
plt.show()
```

5. Transpose the Image

```python
transposed_img = img_cv.T
plt.imshow(transposed_img, cmap='gray')
plt.title("Transposed Image")
plt.axis("off")
plt.show()
```

6. Flip the Image Horizontally and Vertically

```python
h_flip = np.fliplr(img_cv)
v_flip = np.flipud(img_cv)
plt.subplot(1, 2, 1)
plt.imshow(h_flip, cmap='gray')
plt.title("Horizontally")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(v_flip, cmap='gray')
plt.title("Vertically")
plt.axis('off')
plt.show()
```

7. Slice (Crop) a Section of the Image

```python
height, width = img_cv.shape
center_crop = img_cv[height//2-50:height//2+50, width//2-50:width//2+50]

plt.imshow(center_crop, cmap='gray')
plt.axis('off')
plt.show()
```

8. Create a Negative Image

```python
negative_img = 255 - img_cv

plt.imshow(negative_img, cmap='gray')
plt.axis('off')
plt.show()
```

9. Rotate the Image

```python
rotate_img = np.rot90(img_cv, k=1)

plt.imshow(rotate_img, cmap='gray')
plt.axis('off')
plt.show()
```

10. Combine Multiple Transformations

```python
cropped_img = img_cv[height//2-50:height//2+50, width//2-50:width//2+50]
flipped_img = np.fliplr(cropped_img)
brightened_img = flipped_img + 50
clipped_img = np.clip(brightened_img, 0, 255)

plt.imshow(clipped_img, cmap='gray')
plt.axis('off')
plt.show()
```
