```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("new_zealand.jpg")
image.shape
```

Contrast Enhancement

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
hsv_image[:, :, 2] = cv2.equalizeHist(hsv_image[:, :, 2])
contrast_enhanced_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
```

Saturation Enhancement

```python
hsv_image[:, :, 1] = np.clip(hsv_image[:, :, 1] * 1.5, 0, 255)
saturation_enhanced_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(contrast_enhanced_image, cv2.COLOR_BGR2RGB))
plt.title('Contrast Enhanced')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(saturation_enhanced_image, cv2.COLOR_BGR2RGB))
plt.title('Saturation Enhanced')
plt.axis('off')

plt.tight_layout()
plt.show()
```
