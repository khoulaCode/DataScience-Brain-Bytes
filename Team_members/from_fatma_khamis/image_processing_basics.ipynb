```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)
image = np.random.randint(0, 255, (10, 10), dtype=np.uint8)
plt.imshow(image, cmap='gray')
plt.title("Grayscale Image")
plt.colorbar()
plt.show()
```

```python
inversed_img = 255 - image
plt.imshow(inversed_img, cmap='gray')
plt.title("Grayscale Inversion")
plt.colorbar()
plt.show()
```

```python
thresholded_img = np.where(image < 128, 0, 255).astype(np.uint8)
plt.imshow(thresholded_img, cmap='gray')
plt.title("Thresholding")
plt.colorbar()
plt.show()
```

```python
cropped_image = image[3:7, 3:7]
plt.imshow(cropped_image, cmap='gray')
plt.title("Cropping")
plt.colorbar()
plt.show()
```

```python
edges = np.zeros_like(image)
for i in range(image.shape[0]):
    for j in range(image.shape[1] - 1):
        edges[i, j] = image[i, j + 1] - image[i, j]
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.colorbar()
plt.show()
```
