```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"C:\Users\bbuser\maather\data_household\household_power_consumption.txt"
df = pd.read_csv(path,sep=';')
df
```

## pd.to_datetime() conversion

```python
df['comp_data_time'] = df['Date'] + ' ' + df['Time']
df['DateTime'] = pd.to_datetime(df['comp_data_time'], format='%d/%m/%Y %H:%M:%S')
df['DateTime']
```

## Setting datetime index

```python
df.set_index('DateTime', inplace=True)
```

### Resampling (resample('M'), resample('W'))

```python
df.dtypes
```

```python
cols_to_convert = [
    'Global_active_power',
    'Global_reactive_power',
    'Voltage',
    'Global_intensity',
    'Sub_metering_1',
    'Sub_metering_2'
]

# Convert each to numeric, turning errors into NaN
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dtypes
```

```python
daily_avg = df['Global_active_power'].resample('D').mean() # use to get monthly average (resampling:rescale the data to make it more understandable)
daily_avg 
```

```python
weekly_avg = df['Global_active_power'].resample('W').mean()
weekly_avg
```

### Plotting simple trends

```python
fig,axes = plt.subplots(2,1,figsize = (19,25))
daily_avg.plot(ax=axes[0],kind = 'bar',color = 'blue')
axes[0].set_title('Daily Average Global Active Power')
axes[0].grid(True)

weekly_avg.plot(ax=axes[1],kind = 'bar',color = 'red')
axes[1].set_title('Weekly Average Global Active Power')
axes[1].grid(True)
plt.show()



```
