Top 5 countries by average GDP.

```python
df[[' Country ',' Gross Domestic Product (GDP) ']].head(5)
```

Group by Country and calculate:
Mean GDP per capita.

```python
mean_v = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending = False).head(5)
mean_v
```

```python
mean_v.plot(kind = 'bar' , figsize = (8,5))
plt.title(" Top 5 countries ")
plt.ylabel( "Average GDP (USD) ")
plt.xlabel("country")
plt.tight_layout()
plt.show()
```

Countries with the highest inflation in the most recent year.

```python
highest_inflation = df.groupby(' Country ')[' Year '].max().sort_values(ascending = False).head(5)
highest_inflation
```

```python
highest_inflation.plot(kind = 'bar' , figsize = (8,5))
plt.title(" Top 5 countries ")
plt.ylabel( " YEARS ")
plt.xlabel("country")
plt.tight_layout()
plt.show()
```

```python
GDP = df.groupby(' Country ')[' Per capita GNI '].mean().head(10)
GDP
```
