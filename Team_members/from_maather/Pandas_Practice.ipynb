Top 5 countries by average GDP.

```python
df[[' Country ',' Gross Domestic Product (GDP) ']].head(5)
```

Group by Country and calculate:
Mean GDP per capita.

```python
mean_v = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending = False).head(5)
mean_v
```

```python
mean_v.plot(kind = 'bar' , figsize = (8,5))
plt.title(" Top 5 countries ")
plt.ylabel( "Average GDP (USD) ")
plt.xlabel("country")
plt.tight_layout()
plt.show()
```

Countries with the Household consumption expenditure in the most recent year.

```python
latest_year = df[' Year '].max() #to detrmaine last year 
latest_data = df[df[' Year '] == latest_year]# to take the values for last year
final_result = latest_data.groupby(' Country ')[' Household consumption expenditure (including Non-profit institutions serving households) '].sum().sort_values(ascending = False).head(5)
final_result
```

```python
final_result.plot(kind = 'bar' , figsize = (8,5),legend = False)
plt.title(" Top 5 countries ")
plt.ylabel( " Household consumption expenditure ")
plt.xlabel(" Country ")
plt.tight_layout()
plt.show()
```

Max import of goods values


```python
max_goods= df.groupby(' Country ')[' Imports of goods and services '].max().sort_values(ascending = False).head(5)
max_goods
```

```python
max_goods.plot(kind = 'bar' , figsize = (8,5))
plt.title(" Top 5 countries ")
plt.ylabel( " Imports of goods and services ")
plt.xlabel("country")
plt.tight_layout()
plt.show()
```

Mean GDP per capita.

```python
mean_GDP = df.groupby(' Country ')[' Per capita GNI '].mean().head(10)
mean_GDP
```

```python
mean_GDP.plot(kind = 'bar' , figsize = (8,5))
plt.title(" Top 10 countries ")
plt.ylabel( " Per capita GNI ")
plt.xlabel("country")
plt.tight_layout()
plt.show()
```

Total exports over all years.

```python
over_all = df.groupby(' Country ')[' Exports of goods and services '].sum().head(10)
over_all
```

```python
overall_PLOT = over_all.plot(kind = 'bar' , figsize = (8,5))
plt.title(" Top 10 countries ")
plt.ylabel( " Exports of goods and services ")
plt.xlabel("country")
plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize = (19,25))
#plot 1
mean_v.plot(ax=axes[0,0],kind = 'bar',color = 'blue')
axes[0,0].set_title(' Mean GDP per capita')
axes[0,0].grid(True)

# plot 2
final_result.plot(ax=axes[0,1],kind = 'bar',color = 'red')
axes[0,1].set_title('Household consumption expenditure')
axes[0,1].grid(True)

# plot 3

max_goods.plot(ax=axes[1,0],kind = 'bar',color = 'teal')
axes[1,0].set_title("Max import of goods values")
axes[1,0].grid(True)

# plot 4

mean_GDP.plot(ax=axes[1,1],kind = 'bar',color = 'teal')
axes[1,1].set_title('Mean GDP per capita')
axes[1,1].grid(True)
```
