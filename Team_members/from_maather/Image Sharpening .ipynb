```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Load the provided X-ray image.

```python
image = cv2.imread('x-rays.png')
image
```

```python
plt.imshow(image)
plt.show()
```

### Convert to grayscale (if not already).

```python
image_grey = cv2.imread('x-rays.png',cv2.IMREAD_GRAYSCALE)
image_grey
```

```python
plt.imshow(image_grey, cmap = 'grey')
plt.show()
```

```python
image.shape # to check if the picture grey scale or not in this case is not
```

```python
image_grey.shape # to check if the picture grey scale or not in this case it is grey scale
```

### Apply a sharpening kernel (e.g.,[[0,-1,0], [-1,5,-1], [0,-1,0]]).

```python
sharpening=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharped_image = cv2.filter2D(image_grey,-1,sharpening)
sharped_image
```

```python
plt.imshow(sharped_image, cmap = 'grey')
plt.show()
```

### Compare the sharpened image to the original.

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1) 
plt.imshow(image_grey, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(sharped_image, cmap='gray'); plt.title('Sharpened'); plt.axis('off')
plt.tight_layout()
plt.show()
```
