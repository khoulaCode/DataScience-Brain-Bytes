```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
base_url = 'https://www.scrapethissite.com' # given url
inside_card_url = "https://www.scrapethissite.com/pages/frames/?frame=i" 
card_url = 'https://www.scrapethissite.com/pages/frames/?frame=i&family=Carettochelyidae' # link get it from card
current_link = base_url
```

```python
r = requests.get(base_url) # will send a request for get and will save the res into a variable called 'r'
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html') # called a function 'BeautifulSoup' to parse the unreadable 'r.content' to the readable 
print(soup)
```

```python
r = requests.get(inside_card_url) #now for cards 
soup = BeautifulSoup(r.content, 'html')
cards = soup.find_all('div', class_ = "col-md-4 turtle-family-card") #to instract the data from cards(names)
all_cards = {'turtle_name': [], 'description':[]}
for card in cards:
    turtle_tag = card.find('h3', class_ = 'family-name')
    if turtle_tag:
        turtle_names = turtle_tag.text.strip()
        all_cards['turtle_name'].append(turtle_names)
        
    link_tag = card.find('a', href = True)
    if not link_tag:
        description = "No description available"
    else:
        # Follow the link to get description paragraph
        detail_url = base_url + link_tag['href']
        detail_response = requests.get(detail_url)
        detail_soup = BeautifulSoup(detail_response.content, 'html.parser')

        # Get the first <p class="lead"> paragraph
        description_tag = detail_soup.find('p', class_='lead')
        description = description_tag.get_text(strip=True) if description_tag else "No description available"

    # Add to data
    all_cards['description'].append(description)
print(all_cards)
```
