```python
from PIL import Image
# load an image from a file

img_pil = Image.open('download.jpg')
print(f"PIL Image format: {img_pil.format},size: {img_pil.size},mode: {img_pil.mode}")
```

```python
#using matplot to display the RBG converted image
plt.imshow(img_pil)
plt.title('RBG Image')
plt.axis('off')
plt.show
```

```python
import cv2
img_cv = cv2.imread('download.jpg')
img_cv
```

```python
plt.imshow(img_cv)
plt.title('cv2 Image')
plt.axis('off')
plt.show
```

```python
# convert from BGR to RBG 
img_rgb = cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
# convert to grayscale
img_gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_rgb)
plt.title('cv2 img_rgb')
plt.axis('off')
plt.show
```

```python
plt.imshow(img_gray, cmap = 'gray')
plt.title('cv2 img_gray')
plt.axis('off')
plt.show
```

```python
plt.imshow(img_gray[50:,30:], cmap = 'gray')
plt.title('cv2 img_gray')
plt.axis('off')
plt.show
```

```python
plt.imshow(img_gray[100:,50:], cmap = 'gray')
plt.title('cv2 img_gray')
plt.axis('off')
plt.show
```

```python
print("Shape:", img_gray.shape)
print("Data Type:", img_gray.dtype)
print("Min Pixel:", img_gray.min(), "| Max Pixel:", img_gray.max())
```

```python
bright_img = img_gray + 50
plt.imshow(bright_img, cmap='gray')
plt.title("Brightened Image (+50)")
plt.axis('off')
plt.show()
```

```python
bright_img = img_gray + 200
plt.imshow(bright_img, cmap='gray')
plt.title("Brightened +200 (Overflow?)")
plt.axis('off')
plt.show()
```

```python
clipped_img = np.clip(img_gray + 200, 0, 255)
plt.imshow(clipped_img, cmap='gray')
plt.title("Clipped Bright +200")
plt.axis('off')
plt.show()
```

```python
dark_img = np.clip(img_gray - 100, 0, 255)
plt.imshow(dark_img, cmap='gray')
plt.title("Darkened Image (-100)")
plt.axis('off')
plt.show()
```

```python
transposed = img_gray.T
plt.imshow(transposed, cmap='gray')
plt.title("Transposed Image")
plt.axis('off')
plt.show()
```

```python
flip_lr = np.fliplr(img_gray)
flip_ud = np.flipud(img_gray)
```

```python
plt.imshow(flip_lr, cmap='gray')
plt.title("Flipped Left-Right")
plt.axis('off')
plt.show()
```

```python
plt.imshow(flip_ud, cmap='gray')
plt.title("Flipped Up-Down")
plt.axis('off')
plt.show()
```

```python
h, w = img_gray.shape
center_crop = img_gray[h//2 - 50:h//2 + 50, w//2 - 50:w//2 + 50]

plt.imshow(center_crop, cmap='gray')
plt.title("100x100 Center Crop")
plt.axis('off')
plt.show()
```

```python
negative = 255 - img_gray
plt.imshow(negative, cmap='gray')
plt.title("Negative Image")
plt.axis('off')
plt.show()
```

```python
combined = np.clip(np.fliplr(center_crop) + 80, 0, 255)
plt.imshow(combined, cmap='gray')
plt.title("Combined: Crop → Flip → Brighten")
plt.axis('off')
plt.show()
```

```python
downsampled = img_gray[::2, ::2]
neg_resized = cv2.resize(negative, (downsampled.shape[1], downsampled.shape[0]))
flip_resized = cv2.resize(flip_lr, (downsampled.shape[1], downsampled.shape[0]))
crop_resized = cv2.resize(center_crop, (downsampled.shape[1], downsampled.shape[0]))

top = np.hstack((downsampled, neg_resized))
bottom = np.hstack((flip_resized, crop_resized))
collage = np.vstack((top, bottom))

plt.imshow(collage, cmap='gray')
plt.title("2x2 Collage")
plt.axis('off')
plt.show()
```
