```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the X-ray image
# Replace "xray.png" with your file path
img = cv2.imread(r"C:\Users\bbuser\Downloads\1896_x-rays_of_Physikalischer_Verein,_hand_of_man.tif.png")
```

```python
# Step 2: Convert to grayscale (if not already)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Step 3: Define sharpening kernel
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])
```

```python
# Step 4: Apply the sharpening filter
sharpened = cv2.filter2D(gray, -1, sharpen_kernel)
```

```python
# Step 5: Compare original vs sharpened
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original X-ray")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened X-ray")
plt.axis("off")

plt.show()
```

```python
# Step 6: Save the processed image
cv2.imwrite("xray_sharpened.png", sharpened)
```
