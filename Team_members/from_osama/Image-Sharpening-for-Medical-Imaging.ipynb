```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Load the X-ray image

```python
img = cv2.imread("images/1896_x-rays_of_Physikalischer_Verein,_hand_of_man.tif.png")
```

## Step 2: Convert to grayscale

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

## Step 3: Apply sharpening kernel

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

sharpened = cv2.filter2D(gray, -1, kernel)
```

## Step 4: Compare side by side

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(gray, cmap="gray")
plt.title("Original X-ray")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sharpened, cmap="gray")
plt.title("Sharpened X-ray")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## Step 5: Save sharpened image

```python
cv2.imwrite("xray_sharpened.png", sharpened)
```
