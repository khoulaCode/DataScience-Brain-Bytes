## Color Enhancement task

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('walk_view.jpg')
image
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

1. Contrast Enhancement:You should implement histogram equalization. You can perform this on the grayscale version of the image or on the V-channel of the HSV color space to enhance the overall brightness and contrast without affecting the hue.

```python
gray_img = cv2.imread('walk_view.jpg',cv2.IMREAD_GRAYSCALE)
gray_img
```

```python
plt.imshow(gray_img,cmap = 'gray')
plt.show()
```

```python
histogram = cv2.calcHist([image],[0],None,[256],[0,255])
plot_histogram = np.array([c for c in range(histogram.shape[0])])
plt.bar(plot_histogram,histogram[:,0],width = 5)
plt.title('image histogram')
plt.show()
```

2. Saturation Enhancement:You should increase the saturation of the image. This can be done by converting the image to the HSV (Hue, Saturation, Value) color space, multiplying the Saturation channel by a factor (e.g., 1.5), and then converting the image back to RGB. You should handle potential value clipping (values exceeding 255) to avoid unwanted artifacts.

```python
colored_img = cv2.imread('walk_view.jpg',cv2.IMREAD_COLOR)
colored_img
```

```python
HSV_img = cv2.cvtColor(colored_img, cv2.COLOR_BGR2HSV)
plt.imshow(HSV_img)
plt.show()
```

```python
H, S, V = cv2.split(HSV_img)
factor = 1.5
S= np.clip(S.astype(np.float32) * factor, 0, 255).astype(np.uint8)
```

```python
HSV_b = cv2.merge((H, S, V))
RGB_b = cv2.cvtColor(HSV_b, cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(RGB_b)
plt.title("Saturation")
plt.axis("off")
plt.show()
```
