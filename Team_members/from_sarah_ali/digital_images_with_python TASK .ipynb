```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Load and View a Grayscale Image

```python
# Load image in grayscale
img_gray = cv2.imread('view.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# plot
plt.imshow(img_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Check properties
print('Shape:', img_gray.shape)
print('Data type:', img_gray.dtype)
print('Pixel range:', img_gray.min(), 'to', img_gray.max())
```

### Brighten the Image

```python
bright_img = img_gray + 50  # This will overflow!
plt.imshow(bright_img, cmap='gray')
plt.title('Brightened Image (+50)')
plt.axis('off')
plt.show()
```

```python
print('Pixel range:', bright_img.min(), 'to', bright_img.max())
```

### Add a Higher Value

```python
high_bright = img_gray + 300  # Overflow again
clipped = np.clip(img_gray + 300, 0, 255).astype(np.uint8)

plt.imshow(clipped, cmap='gray')
plt.title('Brightened with Clipping (+300)')
plt.axis('off')
plt.show()
```

### Subtract to Darken

```python
dark_img = img_gray - 100  # Underflow!
clipped_dark = np.clip(img_gray - 100, 0, 255).astype(np.uint8)

plt.imshow(clipped_dark, cmap='gray')
plt.title('Darkened Image (-100)')
plt.axis('off')
plt.show()
```

### Transpose the Image

```python
transposed = img_gray.T
plt.imshow(transposed, cmap='gray')
plt.title('Transposed Image')
plt.axis('off')
plt.show()
```

### Flip the Image Horizontally and Vertically

```python
flip_lr = np.fliplr(img_gray)
flip_ud = np.flipud(img_gray)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(flip_lr, cmap='gray')
plt.title('Flipped Left-Right')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(flip_ud, cmap='gray')
plt.title('Flipped Up-Down')
plt.axis('off')
plt.show()
```

### Slice (Crop) a Section of the Image

```python
h, w = img_gray.shape
center_crop = img_gray[h//2 - 50:h//2 + 50, w//2 - 50:w//2 + 50]

plt.imshow(center_crop, cmap='gray')
plt.title('Center Crop (100x100)')
plt.axis('off')
plt.show()
```

### Create a Negative Image

```python
negative_img = 255 - img_gray

plt.imshow(negative_img, cmap='gray')
plt.title('Negative Image')
plt.axis('off')
plt.show()
```

### Rotate the Image

```python
rot90 = np.rot90(img_gray, k=1)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(rot90, cmap='gray')
plt.title('Rotated 90°')
plt.axis('off')
```

```python
rot180 = np.rot90(img_gray, k=2)

plt.subplot(1, 2, 2)
plt.imshow(rot180, cmap='gray')
plt.title('Rotated 180°')
plt.axis('off')
plt.show()
```

### Combine Multiple Transformations

```python
# Crop center
crop = img_gray[h//2 - 50:h//2 + 50, w//2 - 50:w//2 + 50]
# Flip
flipped_crop = np.fliplr(crop)
# Brighten
final = np.clip(flipped_crop + 80, 0, 255).astype(np.uint8)
plt.imshow(final, cmap='gray')
plt.title('Combined Transformations')
plt.axis('off')
plt.show()
```

```python
# Downsample image
downsample = img_gray[h//2 - 50:h//2 + 50, w//2 - 50:w//2 + 50]
# Flip
flipped_downsample = np.fliplr(downsample)
# Brighten
downsample = np.clip(flipped_crop + 80, 0, 255).astype(np.uint8)
plt.imshow(downsample, cmap='gray')
plt.title('Downsample image')
plt.axis('off')
plt.show()
```
