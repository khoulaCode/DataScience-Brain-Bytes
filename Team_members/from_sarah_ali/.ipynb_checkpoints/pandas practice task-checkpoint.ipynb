#### 1. Top 5 countries by average GDP.


used .strip to remove the space

```python
df.columns = df.columns.str.strip()   #remocing space .strip()
```

```python
print(df.columns.tolist())
```

```python
# Group data by Country and calculate mean GDP, then sort descending and take top 5
# Calculate average GDP for each country
avg_gdp = df.groupby('Country')['Gross Domestic Product (GDP)'].mean().sort_values(ascending=False)

# Get top 5
top5_gdp = avg_gdp.head(5)
print("Top 5 countries by average GDP:")
print(top5_gdp)

```

```python
# Plot
# Create bar plot with formatting
import matplotlib.pyplot as plt

top5_gdp.plot(kind='bar', title='Top 5 Countries by Average GDP')
plt.ylabel('Average GDP')
plt.show()
```

# 3. Mean GDP per capita.
*  Calculates GDP per person for each country-year, then finds the 5 countries with highest average GDP per capita over the entire period.

```python
# 3. Top 5 countries by mean GDP per capita
# Create GDP per capita column (GDP divided by population)
df['GDP_per_capita'] = df['Gross Domestic Product (GDP)'] / df['Population']
top5_gdp_per_capita = df.groupby('Country')['GDP_per_capita'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 countries by mean GDP per capita:")
print(top5_gdp_per_capita)

# Visualization (plot)
plt.figure(figsize=(10, 5))
top5_gdp_per_capita.plot(kind='bar', color='lightgreen', title='Top 5 Countries by Mean GDP per Capita (1970-2021)')
plt.ylabel('GDP per Capita (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 4. Max unemployment rate.

*Use changes in inventories as a proxy for economic stress (large changes may indicate economic instability), showing countries with most extreme single-year changes.

```python
# 4. Top 5 countries by max unemployment proxy (using inventory changes)
# Group by country, find maximum inventory changes (proxy for economic stress)
top5_stress = df.groupby('Country')['Changes in inventories'].max().sort_values(ascending=False).head(5)
print("\nTop 5 countries by max economic stress (inventory changes proxy):")
print(top5_stress)

plt.figure(figsize=(10, 5))
top5_stress.plot(kind='bar', color='gold', title='Top 5 Countries by Economic Stress Proxy (1970-2021)')
plt.ylabel('Max Inventory Changes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 5. Total exports over all years.
* Calculates cumulative exports over the entire 51-year period for each country, identifying the 5 biggest exporters.

```python
# 5. Top 5 countries by total exports
# Sum exports across all years for each country, take top 5
top5_exports = df.groupby('Country')['Exports of goods and services'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 countries by total exports (1970-2021):")
print(top5_exports)

#plot
plt.figure(figsize=(10, 5))
top5_exports.plot(kind='bar', color=['orchid',"blue","skyblue","yellow","red"], title='Top 5 Countries by Total Exports (1970-2021)')
plt.ylabel('Total Exports (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
