## Image Sharpening for Medical Imaging

### 1.Import libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### 2. Load the provided X-ray image.

```python
image = cv2.imread('X_RAY_HAND.png')
image
```

```python
plt.imshow(image)
plt.show()
```

### 3. Convert to grayscale (if not already).

```python
grey_IMG = cv2.imread('X_RAY_HAND.png',cv2.IMREAD_GRAYSCALE)
grey_IMG
```

```python
plt.imshow(grey_IMG, cmap = 'grey')
plt.show()
```

```python
image.shape # checking if the picture grey scale or no (here not)
```

```python
grey_IMG.shape # checking if the picture grey scale or no (here yes)
```

### 4. Apply a sharpening kernel (e.g., [[0,-1,0], [-1,5,-1], [0,-1,0]]).

```python
sharpening_kernal=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharp_img = cv2.filter2D(grey_IMG,-1,sharpening_kernal)
sharp_img
```

```python
plt.imshow(sharp_img, cmap = 'grey')
plt.show()
```

### 5. Compare the sharpened image to the original.

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1) 
plt.imshow(grey_IMG, cmap='gray'); plt.title('Original image'); plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(sharp_img, cmap='gray'); plt.title('Sharpened image'); plt.axis('off')
plt.tight_layout()
plt.show()
```
