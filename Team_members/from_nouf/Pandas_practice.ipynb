## Q1: Top 5 countries by average GDP.

```python
df.columns = df.columns.str.strip() # Clean column names
df['Gross Domestic Product (GDP)'].describe()
```

```python
top_five_gdp = df.groupby('Country')['Gross Domestic Product (GDP)'].mean().head(5)
```

```python
# Plot
top_five_gdp.plot(kind='bar', figsize=(10, 5), color='green', title='Mean GDP of First 5 Countries')
plt.ylabel('Mean GDP')
plt.xlabel('Country')
plt.xticks(rotation=35)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Q2: Countries with the Household consumption expenditure in the most recent year.

```python
df.rename(columns={
    'Household consumption expenditure (including Non-profit institutions serving households)': 'household_exp'
}, inplace=True)
```

```python
latest_year = df['Year'].max()
recent_data = df[df['Year'] == latest_year]
recent_household_data = recent_data[['Country', 'Year', 'household_exp']]
recent_household_data
```

```python
# Countries with the Household consumption expenditure in the most recent year.
consumption_exp = recent_household_data.sort_values('household_exp', ascending=False).head(15)
plt.figure(figsize=(14, 6))
plt.bar(consumption_exp['Country'], consumption_exp['household_exp'], color='darkcyan')
plt.title(f'Household consumption expenditure in {latest_year}')
plt.xlabel('Country')
plt.ylabel('Household Expenditure (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## Q3: Group by Country

```python
# this line is used to clean the names of the columns
df.columns = df.columns.str.strip()

# I Grouped by country and then calculated mean , max and sum
summary = df.groupby('Country').agg({
    'Per capita GNI': 'mean',
    'Imports of goods and services': 'max',
    'Exports of goods and services': 'sum'
})
# this line will show the first raws
print(summary.head())
```

### Mean GDP per Capita

```python
top_gdp = summary['Per capita GNI'].sort_values(ascending=False).head(5)

plt.figure(figsize=(6, 4))
plt.bar(top_gdp.index, top_gdp.values, color='skyblue')
plt.title('top countries by Mean GDP per Capita')
plt.xlabel('Country')
plt.ylabel('Per capita GNI ')
plt.xticks(rotation=45, fontsize=8)
plt.grid(axis='y', linestyle='--', alpha=0.5, )
plt.tight_layout()
plt.show()
```

### Max Imports of Goods

```python
top_imports = summary['Imports of goods and services'].sort_values(ascending=False).head(5)

plt.figure(figsize=(6,4))
plt.bar(top_imports.index, top_imports.values, color='salmon')
plt.title('Top 10 Countries by Max Imports of Goods')
plt.xlabel('Country')
plt.ylabel('Imports (USD)')
plt.xticks(rotation=45, fontsize=8)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Total Exports Over All Years

```python
top_exports = summary['Exports of goods and services'].sort_values(ascending=False).head(5)

plt.figure(figsize=(6,4))
plt.bar(top_exports.index, top_exports.values, color='yellow')
plt.title('Top countries by Total Exports Over All Years', fontsize=10)
plt.xlabel('Country', fontsize=8)
plt.ylabel('Exports (USD)', fontsize=8)
plt.xticks(rotation=45, fontsize=8)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
