## Basic insights

#### Top 5 countries by average GDP

```python
average_gdp = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending = False).head(5)
average_gdp
```

```python
plt.figure(figsize=(8,5))
average_gdp.plot(kind="bar", color="skyblue")
plt.title("Top 5 Countries by Average GDP")
plt.xlabel("Country")
plt.ylabel("Average GDP")
plt.tight_layout()
plt.show()
```

#### Countries with the Household consumption expenditure in the most recent year.


```python
Household_consumption = df.groupby(' Country ')[' Household consumption expenditure (including Non-profit institutions serving households) '].max().sort_values(ascending = False).head(10)
Household_consumption
```

```python
Household_consumption.plot(kind='bar', figsize=(8, 5))
plt.title("Household consumption expenditure in the most recent year")
plt.bar(Household_consumption.index,Household_consumption.values,color=['skyblue', 'cornflowerblue', 'goldenrod', 'rosybrown', 'plum'],edgecolor='black')
plt.ylabel("Household consumption expenditure")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

### Group by Country and calculate:

#### Mean GDP per capita.

```python
mean_gdp = df.groupby(' Country ')[' Per capita GNI '].mean().head(10)
mean_gdp
```

```python
plt.figure(figsize=(15, 8))
bars = plt.bar(
    mean_gdp.index,    
    mean_gdp.values,   
    color=['skyblue','lightcoral','lightgreen','gold','plum'],
    edgecolor='black',
    linewidth=0.8
)
for bar in bars:
    yval = bar.get_height()
    xpos = bar.get_x() + bar.get_width()/2
    plt.text(xpos, yval + 0.5, f"{yval:,.0f}",ha='center', va='bottom', fontsize=15)
    
plt.title("Top 10 Countries by Mean GDP per Capita", fontsize = 20)
plt.xlabel("Mean GDP per Capita (USD)",fontsize = 18)
plt.ylabel("Country",fontsize = 18)
plt.tight_layout()
plt.show()
```

#### Max import of goods values

```python
goods_value = df.groupby(' Country ')[' Imports of goods and services '].max().head(5)
goods_value
```

```python
plt.figure(figsize=(20,8))
bars = plt.bar(
    goods_value.index,    
    goods_value.values,   
    color=['skyblue','lightcoral','lightgreen','gold','plum'],
    edgecolor='black',
    linewidth=0.8
)
for bar in bars:
    yval = bar.get_height()
    xpos = bar.get_x() + bar.get_width()/2
    plt.text(xpos, yval + 0.5, f"{yval:,.0f}",ha='center', va='bottom', fontsize=15)
    

plt.title("Max import of goods values", fontsize=14)
plt.xlabel("Imports of goods and services", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.grid()
plt.tight_layout()
plt.show()
```

#### Total exports over all years.

```python
total_exports = df.groupby(' Country ')[' Exports of goods and services '].sum().head(10)
total_exports
```

```python
plt.figure(figsize=(20,8))
bars = plt.bar(
    total_exports.index,    
    total_exports.values,   
    color=['skyblue','lightcoral','lightgreen','gold','plum'],
    edgecolor='black',
    linewidth=0.8
)
for bar in bars:
    yval = bar.get_height()
    xpos = bar.get_x() + bar.get_width()/2
    plt.text(xpos, yval + 0.5, f"{yval:,.0f}",ha='center', va='bottom', fontsize=15)
    
plt.title("Top 10 Countries by Total Exports", fontsize=14)
plt.xlabel("Total Exports (USD)", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.tight_layout()
plt.show()
```
