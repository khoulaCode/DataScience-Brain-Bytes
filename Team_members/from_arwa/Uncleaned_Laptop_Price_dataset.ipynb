```python
# Weight Datatype changed
df.rename(columns={'Weight': 'Weight (kg)'}, inplace=True)
df.head(5)
```

```python
def get_storage_type(memory_str):
    memory_str = memory_str.lower()
    has_ssd = 'ssd' in memory_str
    has_hdd = 'hdd' in memory_str

    if has_ssd and has_hdd:
        return 'SSD + HDD'
    elif has_ssd:
        return 'SSD'
    elif has_hdd:
        return 'HDD'
    else:
        return 'Other'
        
df['StorageType'] = df['Memory'].apply(get_storage_type)
df
```

## Data Analysis

Calculate the average price for each company

```python
avg_company_price = df.groupby('Company')['Price'].mean()
avg_company_price
```

```python
plt.figure(figsize=(20,15))
plt.bar(avg_company_price.index, avg_company_price.values)
plt.title('Average Laptop Price by Brand', fontsize = 16)
plt.xlabel('Company',fontsize = 14)
plt.ylabel('Average Price',fontsize = 14)
plt.show()
```
