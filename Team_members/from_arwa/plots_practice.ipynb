## Scatter plot

```python
np.random.seed(42)
x_scatter = np.random.rand(50) *10
y_scatter = np.random.rand(50) *10 +x_scatter *0.5 + np.random.randn(50)
plt.figure(figsize=(15,8))
#Alpha make the dots / lines more transparent, edgecolors change the outline of the dots
plt.scatter(x_scatter, y_scatter, color='red', alpha=0.6, edgecolors ="black")
plt.title("Scatter Plot of X vs Y", fontsize=16)
plt.xlabel("X values", fontsize=16)
plt.ylabel("Y values", fontsize=16)
plt.legend()
plt.grid(True, linestyle = ':', alpha=0.6)
plt.show()
```

## Bar plots

```python
categories = ['A','B','C','D','E']
values = [23,45,65,18,39]

plt.figure(figsize=(15,8))
plt.bar(categories,values, color = ['skyblue', 'blue', 'goldenrod', 'cornflowerblue', 'plum'])
plt.title("Bar Plot of Category Values", fontsize=16)
plt.xlabel("Category", fontsize=16)
plt.ylabel("Values", fontsize=16)
plt.grid(True, alpha = 0.4)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 65, 18, 39]

plt.figure(figsize=(15, 8))
bars = plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom')
    
plt.title("Bar Plot of Category Values")
plt.xlabel("Category")
plt.ylabel("Values")
plt.grid(True, alpha = 0.4)
plt.show()
```

```python
import math

import matplotlib.pyplot as plt

import matplotlib.colors as mcolors
from matplotlib.patches import Rectangle


def plot_colortable(colors, *, ncols=4, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * ncols + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * ncols)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig
    
plot_colortable(mcolors.CSS4_COLORS)
plt.show()
```

## histogram plot

```python
data_hist = np.random.randn(1000)*15 + 50

plt.figure(figsize=(15,8))
plt.hist(data_hist)
plt.title("Histogram of Data Distribution", fontsize=16)
plt.xlabel("Value", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.show()
```

```python
data_hist = np.random.randn(1000) *15 +50
data_hist_2 = np.random.randn(1000) *15 +30
plt.figure(figsize=(15,8))
plt.grid(True, alpha = 0.6 , linestyle = ':') 
plt.hist(data_hist, bins = 30, color='teal', edgecolor = 'black', alpha = 0.4 )
plt.hist(data_hist_2, bins = 30, color='red', edgecolor = 'black', alpha = 0.4 )
plt.title("Histogram of Data Distribution", fontsize = 16)
plt.xlabel("Values", fontsize = 16)
plt.ylabel("The frequency", fontsize = 16)
plt.savefig('histogramfig.png')
plt.show()
```
