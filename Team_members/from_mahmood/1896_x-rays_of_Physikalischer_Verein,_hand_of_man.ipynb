```python
import cv2 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('1896_x-rays_of_Physikalischer_Verein,_hand_of_man.tif')
```

```python
def plot_image(image_1, image_2,title_1="Orignal", title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot_image(img, img_rgb, 'the bgr', 'the rgb')
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plot_image(img_gray, img_rgb, 'the GRAY', 'the rgb')
```

# Apply a sharpening kernel (e.g.,[[0,-1,0], [-1,5,-1], [0,-1,0]]).

```python
kernel=np.array([[0,-1,0], 
        [-1,5,-1], 
         [0,-1,0]],)
sharped_image =cv2.filter2D(img_gray, -1, kernel)
plot_image(sharped_image, img_gray, 'sharped_image', 'gray image' )
```

```python
img_sharped=cv2.imwrite('sharped_image.jpg', sharped_image)
```
