```python
# Step 1: Ensure column is string
df['square_feet'] = df['square_feet'].astype(str)

# Step 2: Extract number and unit into two new columns
df[['Area_number', 'Area_unit']] = df['square_feet'].str.extract(r'(\d+\.?\d*)\s*(\w+)')

# Step 3: Convert number to float
df['Area_number'] = df['Area_number'].astype(float)
unit_to_sqft = {
    'sqft': 1,
    'sqm': 10.7639,
    'sqyrd': 9.0,
    'acre': 43560,
    'rood': 10890,
    'ground': 2400
}
valid_units = list(unit_to_sqft.keys())
df = df[df['Area_unit'].isin(valid_units)]
df['square_feet'] = df['Area_number'] * df['Area_unit'].map(unit_to_sqft)
df.drop(columns=['Area_number', 'Area_unit'], inplace=True)
df['square_feet'] = df['square_feet'].round(2)
```
