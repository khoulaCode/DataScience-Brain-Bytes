```python
import pandas as pd
```

```python

df = pd.read_json("News_Category_Dataset_v3.json", lines=True)

df.head()
```

```python
keep_categories = ["POLITICS", "TRAVEL", "SPORTS", "HOME & LIVING"]
df_filtered = df[df["category"].isin(keep_categories)]
df_balanced = pd.concat(
    [
        df_filtered[df_filtered["category"] == cat].sample(n=1000, random_state=42)
        for cat in keep_categories
    ],
    ignore_index=True
)
df_balanced
```

```python
df_final = df_balanced[["headline", "category"]].reset_index(drop=True)

df_final.tail()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# limit to top 20 features for simplicity & Create the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the headlines
X = vectorizer.fit_transform(df_final['headline'])

# convert the data to dataframe
tfidf_df = pd.DataFrame(X.toarray(), columns = vectorizer.get_feature_names_out())
tfidf_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
headlines = df_final["headline"].astype(str).tolist()
categories = df_final["category"].astype(str).tolist()

def search(query, top_n=10):
    query_vec = vectorizer.transform([query])
    similarities = cosine_similarity(query_vec, X).flatten() # flatten() we use this to change the vector from 2D TO 1D per headline 
    top_indices = similarities.argsort()[-top_n:][::-1] # to arrange the headlines from more similar to less similar

    results = []
    for i in top_indices:
        results.append({
            "headline": headlines[i],
            "category": categories[i],
            "similarity_score": round(similarities[i], 4)
        })
    
    return results
f_reselt = pd.DataFrame(search("small"))
f_reselt
```
