```python
# Example: assuming the target column is 'score'
target = 'Performance Index'

# Features: all columns except the target
X = df1.drop(columns=[target])

# Target
y = df1[target]

# Check shapes
print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Assuming X = features, y = target
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,   # 20% test set
    train_size=0.8,  # 80% train set
    random_state=42  # reproducible split
)

# Check shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Assume df1 is your dataset and 'Performance Index' is the target
# If 'Extracurricular Activities_Yes' is categorical, convert to 0/1
if df1['Extracurricular Activities_Yes'].dtype == 'bool':
    df1['Extracurricular Activities_Yes'] = df1['Extracurricular Activities_Yes'].astype(int)

# 1️⃣ Separate features and target
X = df1.drop(columns=['Performance Index'])
y = df1['Performance Index']

# 2️⃣ Split dataset into training and testing sets (80/20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    train_size=0.8,
    random_state=42
)

# 3️⃣ Initialize Linear Regression model
model = LinearRegression()

# 4️⃣ Train the model
model.fit(X_train, y_train)

# 5️⃣ Optional: check coefficients
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# 6️⃣ Optional: make predictions on test set
y_pred = model.predict(X_test)

# 7️⃣ Optional: compare first 5 predictions vs actual
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(comparison.head())
```

```python
import pandas as pd

# Get feature names
feature_names = X_train.columns

# Get model coefficients
coefficients = model.coef_

# Combine into a DataFrame for easy interpretation
coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
})

# Sort by absolute impact
coef_df['Abs_Coefficient'] = coef_df['Coefficient'].abs()
coef_df = coef_df.sort_values(by='Abs_Coefficient', ascending=False).drop(columns='Abs_Coefficient')

coef_df
```

```python
coef_df
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# 1️⃣ Generate predictions on the test set
y_pred = model.predict(X_test)

# 2️⃣ Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# 3️⃣ Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# 4️⃣ Calculate R-squared (R²)
r2 = r2_score(y_test, y_pred)

# 5️⃣ Print evaluation metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

```python
from sklearn.linear_model import Ridge, Lasso

ridge_model = Ridge(alpha=0.01)   # alpha = regularization strength
ridge_model.fit(X_train, y_train)

# 5️⃣ Lasso Regression
lasso_model = Lasso(alpha=0.01)  # small alpha to avoid shrinking everything to 0
lasso_model.fit(X_train, y_train)

# 6️⃣ Coefficients and Intercepts
print("🔹 Ridge Regression")
print("Coefficients:\n", pd.Series(ridge_model.coef_, index=X.columns))
print("Intercept:", ridge_model.intercept_)

print("\n🔹 Lasso Regression")
print("Coefficients:\n", pd.Series(lasso_model.coef_, index=X.columns))
print("Intercept:", lasso_model.intercept_)
```

```python
from sklearn.linear_model import SGDRegressor
# Initialize SGDRegressor (with regularization)
sgd_model = SGDRegressor(
    loss='squared_error',   # linear regression loss
    penalty='l1',           # 'l2'=Ridge, 'l1'=Lasso, 'elasticnet'=both
    alpha=0.01,             # regularization strength
    max_iter=1000,
    random_state=42
)

# Train model
sgd_model.fit(X_train, y_train)

# Predictions
y_pred = sgd_model.predict(X_test)

#  Evaluate
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("🔹 SGD Regressor")
print("Coefficients:", sgd_model.coef_)
print("Intercept:", sgd_model.intercept_)
print(f"MSE: {mse:.4f}, RMSE: {rmse:.4f}, R²: {r2:.4f}")
```

```python

# 1️⃣ Features & target
X = df1.drop(columns=['Performance Index'])
y = df1['Performance Index']

# 2️⃣ Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled)

# 3️⃣ Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# 4️⃣ Initialize SGDRegressor (with regularization)
sgd_model = SGDRegressor(
    loss='squared_error',   # linear regression loss
    penalty='l2',           # 'l2'=Ridge, 'l1'=Lasso, 'elasticnet'=both
    alpha=0.01,             # regularization strength
    max_iter=1000,
    random_state=42
)

# 5️⃣ Train model
sgd_model.fit(X_train, y_train)

# 6️⃣ Predictions
y_pred = sgd_model.predict(X_test)

# 7️⃣ Evaluate
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("🔹 SGD Regressor")
print("Coefficients:", sgd_model.coef_)
print("Intercept:", sgd_model.intercept_)
print(f"MSE: {mse:.4f}, RMSE: {rmse:.4f}, R²: {r2:.4f}")
```
