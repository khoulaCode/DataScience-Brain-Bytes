```python
import cv2 as cv2 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests as re
```

```python
img = cv2.imread("fort_image.jpg")
```

```python
def plot_image(image_1, image_2,title_1="Orignal", title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

```python
img_rgb =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt
plot_image(img, img_rgb ,'the original', 'the RGB image')
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plot_image(img_rgb, img_gray, 'the origin', 'the gray Scale')
```

```python
value_otsu,prop_otsu = cv2.threshold(img_gray,0, 255, cv2.THRESH_OTSU)

plot_image(img_rgb, prop_otsu , 'the original', 'the threshod image' )
```

```python
equalized_gray = cv2.equalizeHist(img_gray)

plot_image(img_rgb, equalized_gray , 'the original', 'Etequalized gray' )
```

```python
img_hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plot_image(img_rgb, img_hsv , 'the original', 'HSV Image' )
```

```python
img_hsv2=img_hsv
img_hsv2[:,:,2] = cv2.equalizeHist(img_hsv2[:,:,2])  # Equalize only brightness channel
result = cv2.cvtColor(img_hsv2, cv2.COLOR_HSV2RGB)
plot_image(img_rgb, result , 'the original', ' HSV image' )
```
