```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
economy = pd.read_csv('Global Economy Indicators.csv')
economy
```

```python
economy.columns = economy.columns.str.strip()
```

```python
economy.columns
```

```python
top_gdp = economy.groupby('Country')['Gross Domestic Product (GDP)'].mean().sort_values(ascending=False).head(5)
top_gdp.plot(kind='barh')
plt.show()
```

```python
grouped = economy.groupby('Country').agg(GDP=('Gross Domestic Product (GDP)','mean'),per_capita=('Per capita GNI', 'mean'),total_exports=('Exports of goods and services', 'sum'))
grouped
```
