## Color Enhancement

```python
from PIL import Image, ImageEnhance
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image_path = "New_Zealand_scenic_walk_view.jpg"
img = Image.open(image_path)
```

```python
nz_img=cv2.imread("New_Zealand_scenic_walk_view.jpg")
print(nz_img)
```

```python
plt.imshow(cv2.cvtColor(nz_img,cv2.COLOR_RGB2BGR))
plt.axis("off")
```

```python
# This function will plot two images
def plot_image(image_1, image_2,title_1,title_2):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
    plt.title(title_2)
    plt.show()
```

```python
# Enhance contrast
contrast_enhancer = ImageEnhance.Contrast(img)
img_contrast = contrast_enhancer.enhance(1.5)  # 1.5 = 50% more contrast
```

```python
# Enhance color (saturation)
color_enhancer = ImageEnhance.Color(img_contrast)
img_final = color_enhancer.enhance(1.4)  # 1.4 = 40% more saturation
```

```python
# Show before and after
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(nz_img)
plt.title("Original")
plt.axis("off")

# Enhanced image
plt.subplot(1, 2, 2)
plt.imshow(img_final)
plt.title("Enhanced")
plt.axis("off")

plt.show()
```

```python
# Save and show the result
img_final.save("forest_enhanced.jpg")
img_final.show()
```
