## Image Sharpening for Medical Imaging

```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
import os
cwd = os.getcwd()
cwd 
```

```python
image_path = os.path.join(cwd, "x-ray.png")
```

```python
x_ray=cv2.imread("x-ray.png")
print(x_ray)
```

```python
x_ray = cv2.imread("x-ray.png", cv2.IMREAD_GRAYSCALE)

if x_ray is None:
    raise ValueError("Image not found or path is incorrect")
```

```python
plt.imshow(x_ray, cmap='gray', vmin=0, vmax=255) 
plt.axis('off')
plt.show()
```

```python
def plot_image(image_1, image_2,title_1,title_2):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
    plt.title(title_2)
    plt.show()
```

```python
kernel=np.array([[-1,-1,-1],
                 [-1,9,-1],
                 [-1,-1,-1]])
sharped_image=cv2.filter2D(x_ray, -1,kernel)
plt.imshow(sharped_image,cmap='gray')
plt.show()
```

```python
plot_image(x_ray, sharped_image,"Original Image", "Sharped Image")
```

```python
# Save the processed image
cv2.imwrite("x-ray_sharpened.png", sharped_image)
print("Processed image saved as 'x-ray_sharpened.png'")
```
