##### Total exports over all years.

```python
total_exports = df.groupby(' Country ')[' Exports of goods and services '].sum().sort_values(ascending=False).head(5)

print("Top 5 countries by total exports:", total_exports)
```

```python
#plot
plt.figure(figsize=(10, 5))
total_exports.plot(kind='bar', color= ['darkgreen', 'green', 'forestgreen', 'seagreen', 'limegreen'], title='Top 5 Countries by Total Exports (1970-2021)')
plt.ylabel(' Total Exports (USD) ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
