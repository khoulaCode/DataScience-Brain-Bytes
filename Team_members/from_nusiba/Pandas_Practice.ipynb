### Get basic insights:

#### Top 5 countries by average GDP.

```python
top_5 = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending = False).head(5)
top_5
```

```python
top_5.plot(kind='bar', figsize=(8, 5), color = ['lightblue', 'powderblue', 'cadetblue', 'teal', 'darkslategrey'])
plt.title("Top 5 Countries by Average GDP")
plt.ylabel("Average GDP (USD)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Countries with the Household consumption expenditure in the most recent year.

```python
print(df.columns.tolist())
```

```python
most_recent_year = df[' Year '].max()
Household_consumption_expenditure = df[df[' Year '] == most_recent_year][[' Country ', ' Household consumption expenditure (including Non-profit institutions serving households) ']]
Household_consumption_expenditure
```

### Group by Country and calculate:

```python
df.columns = df.columns.str.strip()

# Check column names
print(df.columns.tolist())
```

### Mean GDP per capita.

```python
mean_gdp = df.groupby(' Country ')[[' Gross Domestic Product (GDP) ']].mean().rename(columns={' Gross Domestic Product (GDP) ': 'Mean GDP per Capita'}).head()

# Plot
mean_gdp.plot(kind='bar', figsize=(8, 5), color= ['maroon', 'firebrick', 'brown', 'indianred', 'lightcoral'], legend=False)
plt.title('Mean GDP per Capita by Country')
plt.ylabel('Mean GDP per Capita')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Max import of goods values

```python
max_imports = df.groupby(' Country ')[[' Imports of goods and services ']].max().rename(columns={'Imports of goods and services': 'Max Imports'}).head()

# Plot
max_imports.plot(kind='bar', figsize=(8, 5), color= ['navy', 'mediumblue', 'royalblue', 'cornflowerblue', 'lightskyblue'], legend=False)
plt.title('Max Imports of Goods and Services by Country')
plt.ylabel('Max Imports')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Total exports over all years.

```python
total_exports = df.groupby(' Country ')[[' Exports of goods and services ' ]].sum().rename(columns={' Exports of goods and services ': 'Total Exports'}).head()

# Plot
total_exports.plot(kind='bar', figsize=(8, 5), color='green', legend=False)
plt.title('Total Exports of Goods and Services by Country')
plt.ylabel('Total Exports')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### This other way to put it all togather (Mean, Max, total) in one plot

```python

grouped = df.groupby(' Country ').agg({
    ' Gross Domestic Product (GDP) ': 'mean',
    ' Imports of goods and services ': 'max',
    ' Exports of goods and services ': 'sum'
}).rename(columns={
    ' Gross Domestic Product (GDP) ': 'Mean GDP per Capita',
    ' Imports of goods and services ': 'Max Imports',
    ' Exports of goods and services ': 'Total Exports'
}).head()  # limit to first 10 countries

grouped
```

```python
grouped.plot(kind='bar', stacked=True, figsize=(8, 5), color=['lightcoral', 'lightskyblue', 'seagreen'])
plt.title('Mean GDP, Max Imports, and Total Exports by Country')
plt.ylabel('Value')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
