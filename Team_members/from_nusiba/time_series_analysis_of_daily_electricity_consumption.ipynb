## Time Series Analysis of Daily Electricity Consumption

### Focus:
* Basic time series handling using pandas
* Convert to datetime, set datetime index
* Perform resampling and slicing

### Objective:
#### Analyze electricity consumption patterns over time using datetime-aware methods. Focus on resampling (daily, weekly, monthly), and slicing time ranges.

```python
import pandas as pd

#upload the features data
path= r"C:\Users\bbuser\Desktop\DataScience-Brain-Bytes\Team_members\from_nusiba\data\household_power_consumption.txt"
df = pd.read_csv(path,sep = ";", low_memory=False)
df.head(10)
```

```python
df.columns
```

### Converting to Datetime:

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')
df.set_index('Datetime', inplace=True)
df.head(10)
```

### Setting Datetime Index:

```python
df.set_index(pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S'), inplace=True)
df.index.name = 'Datetime'
df.head(10)
```

### Time-based slicing 

```python
# Create a longer time series
import numpy as np

# Create a time range for every minute in 2007
date_range = pd.date_range(start='2007-01-01', end='2007-12-31 23:59:00', freq='T')  # 'T' = minute
values = np.random.rand(len(date_range))
df = pd.DataFrame({'Value': values}, index=date_range)

# Slice by year
data_2007 = df.loc['2007']
print(f"Data for 2007: {len(data_2007)} records")

# Slice by month
data_jan_2007 = df.loc['2007-01']
print(f"Data for January 2007: {len(data_jan_2007)} records")

# Slice by specific date range
data_week = df.loc['2007-01-01':'2007-01-07']
print(f"Data for first week of 2007: {len(data_week)} records")

# Slice by time of day (using .between_time())
morning_data = df.between_time('06:00', '12:00')
print(f"Morning data (6 AM - 12 PM): {len(morning_data)} records")
```

### Resampling:

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
# Daily average power
daily_avg = df['Global_active_power'].resample('D').mean()
daily_avg_df = daily_avg.reset_index()
daily_avg_df.columns = ['Date', 'Monthly_Average_Power']
daily_avg_df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
#Plotting 
fig, ax = plt.subplots(figsize=(10, 4))
daily_avg.plot(ax=ax, color='indigo', title='Daily Avg Active Power')

plt.tight_layout()
plt.show()
```

```python
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')
weekly_avg = df['Global_active_power'].resample('W-MON').mean()
weekly_avg_df = weekly_avg.reset_index()
weekly_avg_df.columns = ['Date', 'Weekly_Average_Power']
weekly_avg_df.head(10)
```

```python
#Plotting 
fig, ax = plt.subplots(figsize=(10, 4))
weekly_avg.plot(ax=ax, title='Weekly Avg Active Power', color='maroon')
plt.tight_layout()
plt.show()
```

```python
# Monthly average power
monthly_avg = df['Global_active_power'].resample('M').mean()
monthly_avg_df = daily_avg.reset_index()
monthly_avg_df.columns = ['Date', 'Monthly_Average_Power']
monthly_avg_df.head(10)
```

```python
#Plotting 
fig, ax = plt.subplots(figsize=(10, 4))
monthly_avg.plot(ax=ax, title='Monthly Avg Active Power', color='gold')
plt.tight_layout()
plt.show()
```
