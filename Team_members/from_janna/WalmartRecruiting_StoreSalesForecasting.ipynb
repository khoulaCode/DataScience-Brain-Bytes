### 3. Aggregation with ```groupby()```
### Total Sales by Store

#### Total Sales by Store

````python
sales_by_store = train_df.groupby("Store")["Weekly_Sales"].agg(["sum", "mean", "count"])
sales_by_store.columns = ["Total_Sales", "Avg_Weekly_Sales", "Num_Records"]
sales_by_store = sales_by_store.sort_values("Total_Sales", ascending=False)
sales_by_store.head(10)
````

### Total sales by department

````python
sales_by_dept = train_df.groupby('Dept')['Weekly_Sales'].agg(['sum', 'mean', 'count'])
sales_by_dept.columns = ['Total_Sales', 'Avg_Weekly_Sales', 'Num_Records']
sales_by_dept = sales_by_dept.sort_values('Total_Sales', ascending=False)
sales_by_dept.head(10)
````

### 4. Merging Datasets
### Merge ```train``` with ```stores``` data

````python
train_with_stores = train_df.merge(stores_df, on = "Store", how = "left")
train_with_stores
````

### Merge ```train``` with ```features```

````python
train_with_features = pd.merge(train_df, features_df, on=['Store', 'Date'], how='left')
train_with_features
````

### 5. Cncatenating DataFrames

````python
train_df['Date'] = pd.to_datetime(train_df['Date'])
train_2010 = train_df[train_df['Date'].dt.year == 2010]
train_2011 = train_df[train_df['Date'].dt.year == 2011]
# Show sample vertical concat: (row-wise)
sample1 = train_2010[['Store', 'Dept', 'Weekly_Sales']].head(5)
sample2 = train_2011[['Store', 'Dept', 'Weekly_Sales']].head(5)

concat_rows = pd.concat([sample1, sample2], axis=0)
concat_rows
````

````python
# Show horizontal concat: 
concat_columns = pd.concat([sample1.reset_index(drop=True), sample2.reset_index(drop=True)], axis=1)
concat_columns
````

### 5. Joining DataFrames

````python
stores_indexed = stores_df.set_index("Store")
features_indexed = features_df.set_index("Store")
joined_df = features_indexed.join(stores_indexed, on = "Store")
joined_df.head()
````
