# Time Series Analysis of Daily Electricity Consumption

### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Dataset

```python
df = pd.read_csv(r"C:\Users\bbuser\Downloads\household_power_consumption.txt", sep = ";", low_memory=False)
df.head(10)
```

### DateTime Conversion and Index Setting

```python
df['Datetime'] = df['Date'] + ' ' + df['Time'] # Combine Date and Time
df['Datetime'] = pd.to_datetime(df['Datetime'], format='%d/%m/%Y %H:%M:%S') # Convert to datetime format
df.set_index('Datetime', inplace=True) # Set Datetime as index
df.head(10)
```

### Slicing Time Ranges

```python
# Slice by year
data_2007 = df.loc['2007']
print(f"Data for 2007: {len(data_2007)} records")

# Slice by month
data_jan_2007 = df.loc['2007-01']
print(f"Data for January 2007: {len(data_jan_2007)} records")

# Slice by specific date range
data_week = df.loc['2007-01-01':'2007-01-07']
print(f"Data for first week of 2007: {len(data_week)} records")

# Slice by time of day (using .between_time())
morning_data = df.between_time('06:00', '12:00')
print(f"Morning data (6 AM - 12 PM): {len(morning_data)} records")
```

### Resampling Data

```python
# Convert 'Global_active_power' to numeric
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')
```

```python
# Daily Average
daily_avg = df['Global_active_power'].resample('D').mean()
# Converts to Data Frame
daily_avg_df = daily_avg.reset_index()
daily_avg_df.columns = ['Date', 'Daily_Average_Power']
daily_avg_df.head(10)
```

```python
# Weekly Average
weekly_avg = df['Global_active_power'].resample('W-MON').mean()
weekly_avg_df = weekly_avg.reset_index()
weekly_avg_df.columns = ['Date', 'Weekly_Average_Power']
weekly_avg_df.head(10)
```

```python
# Monthly Average
monthly_avg = df['Global_active_power'].resample('ME').mean()
monthly_avg_df = daily_avg.reset_index()
monthly_avg_df.columns = ['Date', 'Monthly_Average_Power']
monthly_avg_df.head(10)
```

### Data Visualization

```python
fig, axis = plt.subplots(3,1,figsize=(10,12))

daily_avg.plot(ax=axis[0], title='Daily Avg Active Power', color='skyblue')
weekly_avg.plot(ax=axis[1], title='Weekly Avg Active Power', color='coral')
monthly_avg.plot(ax=axis[2], title='Monthly Avg Active Power', color='green')

plt.tight_layout()
plt.show()
```
