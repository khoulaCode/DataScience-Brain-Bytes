### Analyze Key Economic Indicators

#### Calculate the top 5 countries by average GDP

```python
avg_gdp = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending = False).head(5)
avg_gdp_df = avg_gdp.reset_index()
avg_gdp_df
```

```python
plt.figure(figsize=(10, 6))
plt.barh(avg_gdp_df[' Country '], avg_gdp_df[' Gross Domestic Product (GDP) '], color='skyblue')
plt.xlabel("Average GDP (USD)")
plt.title("Top 5 Countries by Average GDP")
plt.tight_layout()
plt.show()
```

#### Countries with household consumption expenditure in most recent year

```python
latest_year = df[df[' Year '] == df[' Year '].max()]
household = latest_year.groupby(' Country ')[' Household consumption expenditure (including Non-profit institutions serving households) '].sum().sort_values(ascending = False).head(5)
household
# Convert to DataFrame
household_df = household.reset_index()
household_df.columns = ['Country', 'Household Consumption (USD)']
household_df
```

```python
plt.figure(figsize = (10, 6))
plt.barh(household_df['Country'], household_df['Household Consumption (USD)'], color = 'mediumseagreen')
plt.xlabel("Household Consumption (USD)")
plt.title("Top 5 Countries by Household Consumption (Most Recent Year)")
plt.tight_layout()
plt.show()
```

#### Grouped country-level metrics

```python
country_metrics = df.groupby(' Country ').agg({
    ' Per capita GNI ': 'mean',
    ' Imports of goods and services ': 'max',
    ' Exports of goods and services ': 'sum'
}).reset_index()
country_metrics
```

```python
# Sort each metric to get top 5 countries (for visualization clarity)
top_gni = country_metrics.sort_values(' Per capita GNI ', ascending=False).head(5)
top_imports = country_metrics.sort_values(' Imports of goods and services ', ascending=False).head(5)
top_exports = country_metrics.sort_values(' Exports of goods and services ', ascending=False).head(5)

# Create 3 rows, 1 column subplot
fig, axs = plt.subplots(3, 1, figsize=(12, 18))
fig.suptitle("Country Metrics Overview", fontsize=18)

# Plot 1: Mean Per capita GNI
axs[0].barh(top_gni[' Country '], top_gni[' Per capita GNI '], color='royalblue')
axs[0].set_title('Top 5 Countries by Mean Per capita GNI')
axs[0].set_xlabel('Mean Per capita GNI (USD)')
axs[0].set_ylabel('Country')

# Plot 2: Max Imports of goods and services
axs[1].barh(top_imports[' Country '], top_imports[' Imports of goods and services '], color='orange')
axs[1].set_title('Top 5 Countries by Max Imports of Goods and Services')
axs[1].set_xlabel('Max Imports (USD)')
axs[1].set_ylabel('Country')

# Plot 3: Total Exports of goods and services
axs[2].barh(top_exports[' Country '], top_exports[' Exports of goods and services '], color='green')
axs[2].set_title('Top 5 Countries by Total Exports of Goods and Services')
axs[2].set_xlabel('Total Exports (USD)')
axs[2].set_ylabel('Country')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
