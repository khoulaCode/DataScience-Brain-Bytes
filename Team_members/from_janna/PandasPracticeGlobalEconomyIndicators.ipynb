### Analyze Key Economic Indicators

#### Calculate the top 5 countries by average GDP

```python
avg_gdp = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending = False).head(5)
avg_gdp
```

```python
avg_gdp.plot(kind='bar', figsize=(8, 5))
plt.title("Top 5 Countries by Average GDP")
plt.ylabel("Average GDP (USD)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Find the 5 countries with the most recent year of data (max year)

```python
highest_inflation = df.groupby(' Country ')[' Year '].max().sort_values(ascending = False).head(5)
highest_inflation
```

```python
plt.figure(figsize=(8, 5)) 
highest_inflation.plot(kind = 'barh')
plt.title("Top 5 Countries by Inflation (Most Recent Year)")
plt.xlabel("IMF Based Exchange Rate")
plt.ylabel("Country")
plt.show()
```

#### Calculate the mean per capita GNI for the top 10 countries

```python
mean_gdp = df.groupby(' Country ')[' Per capita GNI '].mean().head(10)
mean_gdp
```

```python
plt.figure(figsize=(8, 5)) 
mean_gdp.plot(kind = 'barh')
plt.title("Top 10 Countries by Mean GDP per Capita")
plt.xlabel("Mean GDP per Capita (USD)")
plt.ylabel("Country")
plt.show()
```

#### Calculate the total exports of goods and services for the top 10 countries

```python
total_exports = df.groupby(' Country ')[' Exports of goods and services '].sum().head(10)
total_exports
```

```python
plt.figure(figsize=(8, 5))
total_exports.plot(kind = 'bar')
plt.title("Top 10 Countries by Total Exports")
plt.xlabel("Total Exports (USD)")
plt.ylabel("Country")
plt.show()
```
