## Step 2: Data Preprocessing
### 2.1 Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Scale 'Amount' and 'Time' only, rest are already PCA components
df['Amount_scaled'] = StandardScaler().fit_transform(df[['Amount']])
df['Time_scaled'] = StandardScaler().fit_transform(df[['Time']])
df.drop(['Amount', 'Time'], axis=1, inplace=True)
```

### 2.2 Train-Test Split (Stratified)

```python
from sklearn.model_selection import train_test_split

X = df.drop('Class', axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

## Step 3: Baseline Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

# Initialize models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "SVM": SVC(probability=True),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

```python
# Train and store predictions
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    results[name] = {"model": model, "pred": y_pred, "proba": y_proba}
```

## Step 4: Model Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

for name, res in results.items():
    print(f"---- {name} ----")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, res['pred']))
    print("\nClassification Report:")
    print(classification_report(y_test, res['pred'], digits=4))
    print(f"ROC-AUC Score: {roc_auc_score(y_test, res['proba']):.4f}")
    print()
```

## 4.2 ROC Curves

```python
plt.figure(figsize=(10, 6))
for name, res in results.items():
    fpr, tpr, _ = roc_curve(y_test, res['proba'])
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc_score(y_test, res['proba']):.4f})")
    
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid(True)
plt.show()
```

## Step 5: Model Improvements
### 5.1 Resampling Techniques: SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train, y_train)

# Re-train one of the models
rf = RandomForestClassifier()
rf.fit(X_res, y_res)
y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:, 1]

print("After SMOTE:")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf, digits=4))
print(f"ROC-AUC: {roc_auc_score(y_test, y_proba_rf):.4f}")
```

### 5.2 Decision Threshold Tuning

```python
from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba_rf)

plt.figure(figsize=(10, 6))
plt.plot(thresholds, precisions[:-1], 'b--', label='Precision')
plt.plot(thresholds, recalls[:-1], 'g-', label='Recall')
plt.xlabel('Threshold')
plt.title('Precision vs Recall for Different Thresholds')
plt.legend()
plt.grid(True)
plt.show()
```
