# Image Sharpening for Medical Imaging

###  Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Load the X-ray Image

```python
xray_image = cv2.imread("xray.png")
xray_image
```

```python
plt.imshow(xray_image, cmap= "gray")
plt.title("Gray-Scale Image")
plt.axis("off")
plt.show()
```

### Apply Sharpening Kernel

```python
# This function will plot two images
def plot_image(image_1, image_2,title_1,title_2):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1, cmap= "gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2, cmap= "gray")
    plt.title(title_2)
    plt.show()
```

```python
# Define sharpening kernel
sharpening_kernel = np.array([[0, -1,  0],
                              [-1, 5, -1],
                              [0, -1,  0]])

# Apply the sharpening filter
sharpened_image = cv2.filter2D(xray_image, -1, sharpening_kernel)
plot_image(xray_image, sharpened_image, "Original Image", "Sharpened X-ray")
```
