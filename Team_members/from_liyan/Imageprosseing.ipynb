### Image resizing 

```python
import numpy as np
res =  cv2.imread(r"C:\Users\bbuser\OneDrive\Pictures\screen-4.jpg") #loading the image
res1 = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)

#defining scale factors 
scale1 = 3.0
scale2 = 1/3.0

#getting the original dimention 
height, width = res1.shape[:2]

#claculate new diemention
nhiegt = int(height * scale1)
nhiegt = int(width * scale1)

#resizing the image 
zoom = cv2.resize(res1, (nhiegt, nhiegt), interpolation = cv2.INTER_CUBIC)
scaled_image = cv2.resize(res1, (int(width * scale2), int(height * scale2)),interpolation=cv2.INTER_AREA)

#dispaly image
fig, axs = plt.subplots(1, 3, figsize=(10, 4))
axs[0].imshow(res1)
axs[0].set_title('Original Image')
axs[1].imshow(zoom)
axs[1].set_title('Zoomed Image')
axs[2].imshow(scaled_image)
axs[2].set_title('scaled_image')
plt.show()
```

### image rotating 

```python
# Load the image
img = cv2.imread(r"C:\Users\bbuser\OneDrive\Pictures\screen-4.jpg")
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define rotation parameters
center = (image_rgb.shape[1] // 2, image_rgb.shape[0] // 2)
angle = 30
scale = 1

# Create rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)

# Rotate the image
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (img.shape[1], img.shape[0]))

# Display images
fig, axs = plt.subplots(1, 2, figsize=(7, 4))
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(rotated_image)
axs[1].set_title('Rotated Image')
plt.show()
```

### filtering optiontion like blurring, sharping edging 

```python
from PIL import Image, ImageFilter

# Open an image
img = Image.open(r"C:\Users\bbuser\OneDrive\Pictures\screen-4.jpg")

# Apply filters
blur_img = img.filter(ImageFilter.BLUR)
sharp_img = img.filter(ImageFilter.SHARPEN)
edges = img.filter(ImageFilter.FIND_EDGES)

# Convert to NumPy arrays for Matplotlib
original_np = np.array(img)
blur_np = np.array(blur_img)
sharp_np = np.array(sharp_img)
edges_np = np.array(edges)

# Plot all images in one figure
plt.figure(figsize=(15, 5))

# Original
plt.subplot(1, 4, 1)
plt.imshow(original_np)
plt.title('Original')
plt.axis('off')

# Blur
plt.subplot(1, 4, 2)
plt.imshow(blur_np)
plt.title('Blur')
plt.axis('off')

# Sharpen
plt.subplot(1, 4, 3)
plt.imshow(sharp_np)
plt.title('Sharpen')
plt.axis('off')

# Edges
plt.subplot(1, 4, 4)
plt.imshow(edges_np)
plt.title('Edges')
plt.axis('off')

# Show all plots
plt.tight_layout()
plt.show()
```
