### Color Enhancement


###### As a photographer working with a landscape series, I want to automatically enhance the colors and contrast of my photos, So that I can make the images more vivid and visually striking, out the true beauty of the scenery.

### getting the picture

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import PIL as Image
```

```python
forest = cv2.imread(r"C:\Users\bbuser\OneDrive\Pictures\forest.jpg")
```

```python

forest.shape
```

```python
plt.imshow(forest)
plt.title("A picture of a forest")
plt.axis('off')
plt.show()
```

### Histogram equalization

```python
Gray_forest = cv2.cvtColor(forest, cv2.COLOR_BGR2GRAY)
```

```python
equalized_forest = cv2.equalizeHist(Gray_forest)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(Gray_forest, cmap='gray')
plt.title("Original Gray Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_forest, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

plt.show()
```

```python
alpha = 1.5  # Contrast control
beta = 0    # Brightness control
adjusted_forest = cv2.convertScaleAbs(forest, alpha=alpha, beta=beta)

plt.figure(figsize=(10, 5))
```

```python
hsv_forest = cv2.cvtColor(forest, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_forest)
plt.title("HSV Color Space")
plt.axis('off')
plt.show()
```

### Contrast Enhancement:

```python
hsv_forest[:, :, 1] = hsv_forest[:, :, 1] * 1.5  # Increase saturation
equalized_forest = cv2.cvtColor(hsv_forest, cv2.COLOR_HSV2BGR)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(forest, cv2.COLOR_BGR2HSV))
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_forest)
plt.title("Enhanced Image")
plt.axis('off')
```

### Saturation  

```python
factory = 1.5  # Increase saturation by 50%
hsv_forest[:, :, 1] = np.clip(hsv_forest[:, :, 1] * factory, 0, 255)
```

```python
hsv_forest[:, :, 1] = np.clip(hsv_forest[:, :, 1] * factory, 0, 255)

hsv_forest = hsv_forest.astype(np.uint8)
```

```python
saturation_forest = cv2.cvtColor(hsv_forest, cv2.COLOR_HSV2BGR)

plt.figure(figsize=(10, 5)) 
plt.imshow(saturation_forest)
plt.title("Saturation Enhanced Image")
plt.axis('off')
plt.show()
```
