### Image Sharpening for Medical Imaging

#### As a radiologist, I want to sharpen an X-ray image so that fine details in bones and tissues become more visible, helping me make a more accurate diagnosis.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import PIL as Image
```

```python
xray = cv2.imread(r"C:\Users\bbuser\OneDrive\Pictures\xray.png")
```

```python
xray.shape  # Check the shape of the image
```

```python
RGB_xray = cv2.cvtColor(xray, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_xray)
plt.title("Original X-ray Image")
plt.axis('off')
plt.show()  
```

```python
xray.shape  # Check the shape of the image
```

### 2.CONVET TO GRAYSCALE

```python
gray_xray = cv2.cvtColor(xray, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_xray, cmap='gray')
plt.title("Grayscale X-ray Image")
plt.axis('off')
plt.show()  
```

### 3.sharppinging

```python
kernal = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
sharpen_xray = cv2.filter2D(gray_xray, -1, kernal)
```

```python
def plot_image(sharpen_xray, gray_xray, title1, title2):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(sharpen_xray, cmap='gray')
    plt.title(title1)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(gray_xray, cmap='gray')
    plt.title(title2)
    plt.axis('off') 

    plt.show()
```

```python
plot_image(sharpen_xray, gray_xray, "Sharpened X-ray Image", "Grayscale X-ray Image")
```

### 5.save the processed image 

```python
cv2.imwrite('sharpened_xray.png', gray_xray)
```
