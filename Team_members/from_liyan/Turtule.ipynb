```python
import requests
from bs4 import BeautifulSoup

def get_turtles():
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    # Step 1: Load the main page
    base_url = "https://www.scrapethissite.com/pages/frames/"
    response = requests.get(base_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Step 2: Find iframe and get its src
    iframe = soup.find("iframe")
    iframe_src = iframe["src"]

    # Step 3: Load the iframe content
    iframe_url = "https://www.scrapethissite.com" + iframe_src
    iframe_response = requests.get(iframe_url, headers=headers)
    iframe_soup = BeautifulSoup(iframe_response.text, 'html.parser')

    # Step 4: Select turtle cards
    turtle_cards = iframe_soup.select('div.turtle-family-card')

    # Step 5: Extract and filter turtle names
    turtle_names = []
    for card in turtle_cards:
        name_tag = card.select_one("h3.common-name")
        if name_tag:
            name = name_tag.text.strip()
            turtle_names.append(name)

    return turtle_names

# Run and print
turtles = get_turtles()
print(turtles)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def get_family_details(details_url):
    response = requests.get(details_url, headers={"User-Agent": "Mozilla/5.0"}) #This sends an HTTP GET request to the URL stored in details_url. Basically, it asks that web page to send its content back.
    soup = BeautifulSoup(response.text, "html.parser")

    details = {
        "nickname": None,
        "year_discovered": None,
        "discoverer": None,
        "description": None
    }
# details is a created dictinoarty where we have all the elements we need 
    p_lead = soup.find('p', class_='lead')
    if p_lead:
        text = p_lead.get_text(strip=True)
        details["description"] = text
# This looks for the first <p> tag with the class lead in the HTML parsed by BeautifulSoup.
        # Extract nickname (inside <strong class="common-name">)
        strong_tag = p_lead.find('strong', class_='common-name')
        if strong_tag:
            details["nickname"] = strong_tag.text.strip()

        # Extract year discovered (4 digit number)
        year_match = re.search(r'(\d{4})', text)
        if year_match:
            details["year_discovered"] = year_match.group(1)

        # Extract discoverer (after 'by' till the period or end)
        discoverer_match = re.search(r'by ([A-Za-z\s]+)[\.\,]?', text)
        if discoverer_match:
            details["discoverer"] = discoverer_match.group(1).strip()

    return details

# Example usage:
url = "https://www.scrapethissite.com/pages/frames/?frame=i&family=Carettochelyidae"
result = get_family_details(url)
print(result) # okay so the problem is the details are inside a button called read more inside (/html/body/div/div/p /html/body/div/div/p<p class="lead">
                        #The <span class="family-name">Carettochelyidae</span> family of turtles — more commonly known as "<strong class="common-name">Pig-nosed turtle</strong>" — were first discovered in 1887 by Boulenger.
                    #</p>)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_family_cards(iframe_url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(iframe_url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    cards = []
    for div in soup.select("div.turtle-family-card"): # here we looped through family card
        family_name = div.select_one("h3.family-name").text.strip() #here we extracted the family name
        href = div.select_one("a.btn")["href"] # here we Extract the href URL from the "Learn more" button becuse the details are inside btn
        family_name = div.select_one("h3.family-name").text.strip()
        href = div.select_one("a.btn")["href"]
        details_url = "https://www.scrapethissite.com" + href #Build the full URL for details
        cards.append({"family_name": family_name, "details_url": details_url}) #Add a dictionary with family name and details URL to a list
    return cards
        details_url = "https://www.scrapethissite.com" + href
        cards.append({"family_name": family_name, "details_url": details_url})
    return cards
#we return all the cards :)
def get_family_details(details_url): # This function takes one argument, details_url, which is the URL of a turtle family’s detailed info page
    headers = {"User-Agent": "Mozilla/5.0"} #Set up headers and request the page
    response = requests.get(details_url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser") #Parse the HTML

    p_tag = soup.select_one("p.lead") #Find the <p> tag with class lead
    if not p_tag: #If no such <p> tag is found
        return {"nickname": None, "year_discovered": None, "discoverer": None, "description": None}

    text = p_tag.get_text(separator=" ", strip=True) #Get the text content inside the <p> tag

    # Extract nickname inside quotes after 'known as'
    import re #Import re module for regex matching
#Extract the nickname
    nickname_match = re.search(r'known as\s+"([^"]+)"', text)
    nickname = nickname_match.group(1) if nickname_match else None
#Looks for the phrase known as "some nickname" in the text.

#r'known as\s+"([^"]+)"' means:

#known as followed by whitespace \s+

#followed by "..." quotes capturing everything inside except quotes ([^"]+)

#If found, extract the nickname inside the quotes; otherwise, None
#-------------------------------------------------------------------------------
    # Extract year discovered (4 digits)
    year_match = re.search(r'first discovered in (\d{4})', text)
    year = year_match.group(1) if year_match else None
#Searches for a 4-digit number after the phrase first discovered in.

#Captures the year like "1887".

#Returns year as a string, or None if not found.
#---------------------------------------------------------------------------------
    # Extract discoverer (last word before period after year)
    discoverer_match = re.search(r'discovered in \d{4} by (\w+)', text)
    discoverer = discoverer_match.group(1) if discoverer_match else None
#Searches for the discoverer's name after the phrase discovered in [year] by [name].

#(\w+) captures a word (letters only) right after by.

#Returns discoverer's name as string or None.
#--------------------------------------------------------------------------------
    return {
        "nickname": nickname,
        "year_discovered": year,
        "discoverer": discoverer,
        "description": text
    }
#Return a dictionary with all extracted details
#------------------------------------------------------------------
def scrape_all_turtles():
    iframe_url = "https://www.scrapethissite.com/pages/frames/?frame=i"
    families = get_family_cards(iframe_url)
    for fam in families:
        details = get_family_details(fam["details_url"])
        fam.update(details)
    return families

# Run and create dataframe
turtle_data = scrape_all_turtles()
df = pd.DataFrame(turtle_data)

df
```
