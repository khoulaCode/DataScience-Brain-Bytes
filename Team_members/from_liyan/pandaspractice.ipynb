# Top 5 countries by average GDP

```python
av_gdp = df.groupby('Country')['Gross Domestic Product (GDP)'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(15, 8))
categories = av_gdp.index
values = av_gdp.values

bars = plt.bar(categories, values, color=["purple", "red", "skyblue", "yellow", "green"] * 2)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, round(yval, 2),
             ha='center', va='bottom')

plt.title("Top 5 countries by average GDP")
plt.xlabel("Country")
plt.ylabel("average GDP")
plt.grid(True, alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Countries by Avg Household Consumption Expenditure

```python
import matplotlib.pyplot as plt

consumption_avg = df.groupby('Country')['Household consumption expenditure (including Non-profit institutions serving households)'].max().sort_values(ascending=False).head(5)

plt.figure(figsize=(15, 8))
categories = consumption_avg.index
values = consumption_avg.values

bars = plt.bar(categories, values, color=["purple", "red", "skyblue", "yellow", "green"] * 2)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, round(yval, 2),
             ha='center', va='bottom')

plt.title("Countries by Avg Household Consumption Expenditure")
plt.xlabel("Country")
plt.ylabel("Avg Expenditure")
plt.grid(True, alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Total exports over all years

```python
total_export = df.groupby('Country')['Exports of goods and services'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(15, 8))
categories = total_export.index
values = total_export.values

bars = plt.bar(categories, values, color=["maroon", "hotpink", "navy", "plum", "cyan"] * 2)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, round(yval, 2),
             ha='center', va='bottom')

plt.title("Total exports over all years.")
plt.xlabel("Country")
plt.ylabel("Avg Expenditure")
plt.grid(True, alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Mean GDP per capita.

```python
df['GDP per capita'] = df['Gross Domestic Product (GDP)'] / df['Population']
mean_gdp = df.groupby('Country')['GDP per capita'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(15, 8))
categories = mean_gdp.index
values = mean_gdp.values

bars = plt.bar(categories, values, color=["lightgreen", "green", "coral", "lime", "tan"] * 2)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, round(yval, 2),
             ha='center', va='bottom')

plt.title("Mean GDP per capita.")
plt.xlabel("Country")
plt.ylabel("Mean GDP per capita")
plt.grid(True, alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
