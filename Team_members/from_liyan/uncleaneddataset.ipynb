```python
def get_storage_type(memory_str):
    memory_str = memory_str.lower()
    has_ssd = 'ssd' in memory_str
    has_hdd = 'hdd' in memory_str

    if has_ssd and has_hdd:
        return 'SSD + HDD'
    elif has_ssd:
        return 'SSD'
    elif has_hdd:
        return 'HDD'
    else:
        return 'Other'
        
df['StorageType'] = df['Memory'].apply(get_storage_type)
df
```

```python
def get_core_intel(cpu):
    cpu = cpu.lower()
    has_i7 = 'i7' in cpu
    has_i5 = 'i5' in cpu
    has_i3 = 'i3' in cpu

    if has_i7 and has_i5 and has_i3:
        return 'i7 + i5 + i3'
    elif has_i7:
        return 'i7'
    elif has_i5:
        return 'i5'
    elif has_i3:
        return 'i3'
    else:
        return 'Other'
        
df['CoreIntel'] = df['Cpu'].apply(get_core_intel)
df
```

```python
def toach_screen(ScreenResolution_1):
    ScreenResolution_1 = str(ScreenResolution_1).lower()
    
    if 'touchscreen' in ScreenResolution_1:
        return 'Yes'
    else:
        return 'No'

df['Touchscreen'] = df['ScreenResolution'].apply(toach_screen)
df
```

# Train model

```python
# Separate features and target
X = df.drop('Price', axis=1)
y = df['Price']

# One-hot encode categorical columns
X_encoded = pd.get_dummies(X, columns=[
    'Company', 'TypeName', 'ScreenResolution', 'Cpu', 'Memory', 'Gpu',
    'OpSys', 'StorageType', 'CoreIntel', 'Touchscreen'
], drop_first=True)  # drop_first=True avoids multicollinearity

# Split the dataset into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# Train the Linear Regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse:.2f}')
print(f'R2 score: {r2:.2f}')
```
