# join()

```python
# Clean column names becuse we kept getting an error on the word store 
df2.columns = df2.columns.str.strip()

# Reset index to aviod the mistsake
if df2.index.name == 'Store':
    df2.reset_index(inplace=True)

# Print to confirm that store is going to match 
print(df2.columns)

# the actual process of merging we merged df1 and df2 based on the common value 'store' we can see the both tables are merged on the left
merged_df = pd.merge(df1, df2, on='Store', how='left')
merged_df
```
