#### Calculate the average GDP for each country, then extract the top 5 countries in this average.

```python
print(df.columns.tolist())
```

```python
top_gdp = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].mean().sort_values(ascending=False).head(5)
print("Top 5 countries by average GDP:\n", top_gdp)
```

 #### I use {.sort_values(ascending=False)} to arrange the values from highest to lowest. After that i use {.head(5)} that will take frist 5 country

# point 2 -
### The countries suffering from the highest inflation rates in the past year, we need to determine:
#### -The latest year in the data.
#### -Countries with the Household consumption expenditure in the most recent year.
#### -Present the top 5 countries in terms of Countries with the Household for that year.

```python
df.columns = df.columns.str.strip()


latest_year = df['Year'].max()


df_latest = df[df['Year'] == latest_year]


top5 = df_latest[['Country', 'Household consumption expenditure (including Non-profit institutions serving households)']] \
        .sort_values(by='Household consumption expenditure (including Non-profit institutions serving households)', ascending=False) \
        .head(5)

print(top5)
```

 -->  To identify the top 5 countries with the highest household consumption expenditure in the most recent year. It begins by using df.columns.str.strip() to remove extra spaces from column names, ensuring clean and error-free access. The latest_year = df['Year'].max() command finds the most recent year in the dataset. Then, df[df['Year'] == latest_year] filters the DataFrame to include only records from that year. The code selects the relevant columns (Country and Household consumption expenditure) and uses .sort_values(ascending=False).head(5) to sort the countries by expenditure in descending order and retrieve the top 5. This allows for a focused analysis of recent consumer spending trends across countries.

# Point 3:

```python
import pandas as pd

df.columns = df.columns.str.strip()

grouped_stats = df.groupby('Country').agg({
    'Gross Domestic Product (GDP)': 'mean',
    'Imports of goods and services': 'max',
    'Exports of goods and services': 'sum'
}).reset_index()


print(grouped_stats.head(5))
```

# Point 4:
### Top 5 countries by Average GDP

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.barh(top_gdp.index, top_gdp.values) 
plt.gca().invert_yaxis()  
plt.title("Top 5 Countries by Average GDP")
plt.xlabel("Average GDP")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

# point 5
#### Top 5 countries by  Household consumption

```python
plt.figure(figsize=(10, 5))
plt.barh(top5['Country'], top5['Household consumption expenditure (including Non-profit institutions serving households)'])
plt.gca().invert_yaxis()
plt.title(f"Top 5 Countries by Household Consumption Expenditure ({latest_year})")
plt.xlabel("Household Consumption Expenditure")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

## Top 5 Countries by Maximum Imports

```python
top_imports = grouped_stats.sort_values(by='Imports of goods and services', ascending=False).head(5)

plt.figure(figsize=(10, 5))
plt.barh(top_imports['Country'], top_imports['Imports of goods and services'])
plt.gca().invert_yaxis()
plt.title("Top 5 Countries by Maximum Imports")
plt.xlabel("Max Imports")
plt.tight_layout()
plt.show()
```

## Top 5 Countries by Total Exports

```python
top_exports = grouped_stats.sort_values(by='Exports of goods and services', ascending=False).head(5)

plt.figure(figsize=(10, 5))
plt.barh(top_exports['Country'], top_exports['Exports of goods and services'])
plt.gca().invert_yaxis()
plt.title("Top 5 Countries by Total Exports")
plt.xlabel("Total Exports")
plt.tight_layout()
plt.show()
```

To reach bar charts we create it by using Pandas for data aggregation and Matplotlib for visualization. The command df.groupby('Country').agg({...}) groups the dataset by country and calculates the average GDP, maximum imports, and total exports for each country. The .sort_values(...).head(5) command then selects the top 5 countries based on each metric. The chart is plotted using plt.barh(...), which creates a horizontal bar chart, and plt.gca().invert_yaxis() is used to display the highest values at the top. Finally, plt.tight_layout() is called to automatically adjust the spacing between chart elements, ensuring that titles, labels, and bars are displayed clearly without overlapping or being cut off.
