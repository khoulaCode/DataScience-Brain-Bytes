```python
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
y= np.sin(x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,6))
plt.plot(x,y, color = 'red', linestyle = '--' ) # we can use (-.), (:), (--)
plt.title("Sine and cosine waves", fontsize = 10)
plt.legend()
plt.xlabel("X-axis", fontsize = 10)
plt.ylabel("Y-axis", fontsize = 10)
plt.grid(True)
plt.show()
```

# histogram

```python
data_hist = np.random.randn(1000)*15+50
data_hist2 = np.random.randn(1000)*15+30


plt.figure(figsize = (15,8))
plt.grid(True, alpha =0.3, linestyle=":")
plt.hist(data_hist, bins =10,  color = 'teal', edgecolor = 'black', alpha = 0.4)
plt.hist(data_hist2, bins =10,  color = 'red', edgecolor = 'black', alpha = 0.4)
plt.title('hist of data distribution')
plt.xlabel('value')
plt.ylabel('frequency')
plt.savefig("figure1.png") # we can save it as pdf or png
plt.show()
```

# Boxplot

```python
data_to_plot = [data_hist,data_hist2]
label = ['Group 1','Group 2']

plt.figure(figsize = (15, 8))
plt.boxplot(data_to_plot,patch_artist =True, labels = label, boxprops =dict(facecolor= 'lightgreen'))
plt.title("Box plot of data group")
plt.xlabel("Group")
plt.ylabel("Value")
plt.grib(axis="y", linestyle = "--", alpha = 0.6)
```
