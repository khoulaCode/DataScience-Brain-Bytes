```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("Data/household_power_consumption.txt", sep=";")
df
```

## 1-Loading and Preparing the Datetime Data ..
#### The dataset initially contained separate Date and Time columns. To perform time series operations efficiently, I combined these two into a new column called {Datetime}, and then converted it into a proper datetime object using:

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], dayfirst=True, errors='coerce')
df['Datetime']
```

##### --> A new column named Datetime is created, combining the original Date and Time columns into a single datetime format.No visible output, but now each row has a proper timestamp. 

```python
df.columns
```

### 2- Drop Invalid Dates and Set as Index

```python
df = df.dropna(subset=['Datetime'])  
df.set_index('Datetime', inplace=True)  
df.drop(columns=['Date', 'Time'], inplace=True)  
```

#### --> Still no visible output, but this step does 3 things:

##### Removes rows with invalid or missing datetime values.

##### Sets Datetime as the index of the DataFrame.

##### Deletes the original Date and Time columns (since they're now redundant).

#### 3-  Convert All Columns to Numeric

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

##### --> No visible output. This loop converts all the string-based columns (like 'Global_active_power', 'Voltage', etc.) into numeric values. Invalid entries (like '?') become NaN. This ensures all data is numeric for calculations and plotting.

#### 4-  Plot the Raw Power Usage

```python
df['Global_active_power'].plot(figsize=(15, 5), title='Global Active Power (Raw)',color='red')
plt.ylabel('kW')
plt.xlabel('Datetime')
plt.show()
```

#### This generates a line chart showing raw power consumption over time Because, the data is in high frequency (possibly minute-level), the chart might look noisy but shows general usage spikes.

X-axis: datetime

Y-axis: power consumption in kilowatts (kW)

Title: "Global Active Power (Raw)"

#### 5- Daily Resampling

```python
df_daily = df['Global_active_power'].resample('D').mean()
df_daily.plot(figsize=(15,5), title='Daily Average Global Active Power',
    color='gray')
plt.ylabel('kW')
plt.show()
```

#### Creates and displays a line chart of daily average electricity usage.

X-axis: each day

Y-axis: average power used that day

Result: smoother than the raw data â€” easier to see trends

This helps detect weekly cycles, weekends, or specific high-usage days.

### Weekly Resampling

```python
df_weekly = df['Global_active_power'].resample('W').mean()
df_weekly.plot(figsize=(15,5), title='Weekly Average Global Active Power',
    color='purple')
plt.ylabel('kW')
plt.show()
```

### Displays a line chart of weekly average power consumption.

Smoother than daily

Ideal for detecting mid-term patterns like weekly usage trends, drops during holidays, etc.

### Monthly Resampling

```python
df_monthly = df['Global_active_power'].resample('M').mean()
df_monthly.plot(figsize=(15,5), title='Monthly Average Global Active Power',
    color='orange')
plt.ylabel('kW')
plt.show()

```

### Generates a line plot showing monthly averages.

Very smooth trend

Reveals seasonal patterns â€” e.g., more electricity in summer or winter

Helps in long-term forecasting or comparing month-over-month usage

### 6- Slice January 2007  by using loc[]

```python
jan_2007 = df.loc['2007-01']
jan_2007['Global_active_power'].plot(figsize=(15,5), title='January 2007 - Global Active Power',
    color='green')
plt.ylabel('kW')
plt.show()
```

### --> This extracts and plots data only for January 2007.

X-axis: days of Jan 2007

Y-axis: actual power values

Helps identify usage patterns within a specific month

### Slice December 20â€“27, 2009

```python
week_sample = df.loc['2009-12-20':'2009-12-27']
week_sample['Global_active_power'].plot(figsize=(15,5), title='Week Sample - Dec 2009',
    color='black')
plt.ylabel('kW')
plt.show()
```

#### --> Plots power usage from Dec 20 to Dec 27, 2009.

Helpful to zoom into specific events (e.g., Christmas week)

Great for analyzing unusual patterns during holidays
