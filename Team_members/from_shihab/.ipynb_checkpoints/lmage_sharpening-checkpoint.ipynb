# Image Sharpening for Medical Imaging

### As a radiologist, I want to sharpen an X-ray image so that fine details in bones and tissues become more visible, helping me make a more accurate diagnosis.

```python
import cv2
from IPython.display import Image, display
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path= "C:Imges\sharpening.png"
```

# Load the provided X-ray image.

```python
img_bgr = cv2.imread(img_path)

gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY) if img_bgr.ndim == 3 else img_bgr.copy()
gray.shape, gray.dtype
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original X-ray Image")
plt.axis("off")
plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]], dtype=np.float32)

sharpened = cv2.filter2D(gray, ddepth=-1, kernel=kernel)
sharpened = np.clip(sharpened, 0, 255).astype("uint8")
```

```python
plt.figure(figsize=(6,6))
plt.imshow(sharpened, cmap='gray')  
plt.title("Sharpened (shown)")
plt.axis('off')
plt.show()
```

```python
# Show the original in RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(img_rgb)
plt.title("Original X-ray Image")
plt.axis("off")
plt.show()

# Convert original to grayscale for fair comparison
plt.figure(figsize=(3,3))
plt.imshow(sharpened, cmap='gray')  
plt.title("Sharpened (shown)")
plt.axis('off')
plt.show()
```

```python
kernel_stronger = np.array([[0, -1, 0],
                            [-1, 9, -1],
                            [0, -1, 0]], dtype=np.float32)
sharpened2 = cv2.filter2D(gray, -1, kernel_stronger)
sharpened2 = np.clip(sharpened2, 0, 255).astype('uint8')

plt.figure(figsize=(10,5))
plt.imshow(np.hstack([gray, sharpened2]), cmap='gray')
plt.title("Left: Original  |  Right: Stronger Kernel (center=9)")
plt.axis('off')
plt.show()
```

```python
cv2.imwrite("xray_sharpened_strong.png", sharpened2)
```
