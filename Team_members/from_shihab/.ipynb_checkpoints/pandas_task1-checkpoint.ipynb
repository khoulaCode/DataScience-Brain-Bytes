# point 2 -
### The countries suffering from the highest inflation rates in the past year, we need to determine:
#### -The latest year in the data.
#### -Extract the inflation value for that year only.
#### -Rank the countries according to the highest inflation rate.
#### -Present the top 5 countries in terms of inflation for that year.

```python
df = df.sort_values([' Country ', ' Year '])

#1
df[' GDP_inflation_rate '] = df.groupby(' Country ')[' Gross Domestic Product (GDP) '].pct_change() * 100

#2
most_recent_year = df[' Year '].max()
#3
top_inflation_calc = df[df[' Year '] == most_recent_year][[' Country ', ' GDP_inflation_rate ']].dropna().sort_values(by=' GDP_inflation_rate ', ascending=False).head(5)

print(f"Top 5 countries by estimated GDP-based inflation in {most_recent_year}:\n", top_inflation_calc)
```

##### -> Due to the absence of a direct column representing the inflation rate in the table, we calculated the estimated inflation using the relative change in the [Gross Domestic Product (GDP)] for each country from year to year. lets explain step by step:

##### 1- used the [pct_change()] function to calculate the annual percentage change in the GDP of each country. This calculates the change as a percentage between a year and its previous year within each group of countries individually.

##### 2- use most_recent_year = df['Year'].max() to know the latest year in the data.
##### 3-I extracted the top 5 countries according to the derived inflation rate for this year.

### This method does not reflect the real inflation associated with consumer prices (CPI), but it is a good estimative method using the change in GDP.

```python
import pandas as pd

df.columns = df.columns.str.strip()


unemp_df = pd.DataFrame({
    'Country': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'Unemployment': [11.2, 13.5, 9.8, 4.5, 10.0]
})


unemp_df.columns = unemp_df.columns.str.strip()




df = pd.merge(df, unemp_df, on='Country', how='left')


grouped_stats = df.groupby('Country').agg({
    'Per capita GNI': 'mean',
    'Unemployment': 'max',
    'Exports of goods and services': 'sum'
}).reset_index()


print(grouped_stats.head(10))
```
