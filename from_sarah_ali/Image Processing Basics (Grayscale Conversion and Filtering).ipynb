## Image Processing Basics (Grayscale Conversion and Filtering)
#### Scenario: You're working on a basic image processing application. You're given a simple grayscale image represented as a NumPy array where each pixel value indicates intensity (0 for black, 255 for white). You need to perform some fundamental operations.

```python
import numpy as np
```

1. Image Representation
* Create a 2D NumPy array (e.g., 10x10) to represent a grayscale image. Initialize it with random integer pixel values between 0 and 255.

```python
image = np.random.randint(0, 256, size=(10, 10), dtype=np.uint8)
#(np.uint8) Best for grayscale images, compact and fast
#stands for Unsigned Integer 8-bit, which means:Values range from 0 to 255 (ideal for grayscale images).#Each pixel is stored using 1 byte (8 bits).
print("Original Image:\n", image)
```

2. Display (Conceptual): 
* Describe how you would conceptually "display" this image (e.g., using a library like Matplotlib, though you don't need to actually plot it unless you wish to).

```python
import matplotlib.pyplot as plt

plt.imshow(image, cmap='gray', vmin=0, vmax=255)   # used VMIN , VMAX for scaling from black (from black to white)
#imshow = image show
#plt.imshow(image)Displays the 2D NumPy array as an image
#	Uses grayscale coloring (0 = black, 255 = white)
plt.colorbar()  
#.colorbar() color scale bar on the side (right)
plt.title("Grayscale Image orignal")
plt.show()
```

3. Grayscale Inversion: 
* Invert the image (black becomes white, white becomes black). For a pixel value p, the inverted value is 255 - p.

```python
#first orignal
image = np.random.randint(0, 256, size=(10, 10), dtype=np.uint8)
print("Original Image:\n", image)

#inverted imag
inverted_image = 255 - image 

print("inverted Image:\n", inverted_image)


plt.imshow(inverted_image, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.title("Inverted Image")
plt.show()
```

4. Thresholding: 

* Apply a simple thresholding operation: all pixels below a certain value (e.g., 128) become 0 (black), and all pixels equal to or above become 255 (white).

Convert the grayscale image into a binary image (only black and white) using a threshold


```python
#first orignal
image = np.random.randint(0, 256, size=(10, 10), dtype=np.uint8)
print("Original Image:\n", image)

#thresholded image
thresholded_image = np.where(image < 128, 0, 255).astype(np.uint8)
print("thresholded image:\n",thresholded_image)

plt.imshow(thresholded_image, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.title("Thresholded Image")
plt.show()
```

5. Cropping: 
 
* Extract a smaller "region of interest" (e.g., a 4x4 sub-image from the center) using array slicing.

```python
#first orignal
image = np.random.randint(0, 256, size=(10, 10), dtype=np.uint8)
print("Original Image:\n", image)
#cropped
cropped_image = image[3:7, 3:7]
# NumPy slicing uses [row_start:row_end, col_start:col_end].
#Rows: 3, 4, 5, 6 → total 4 rows
#Columns: 3, 4, 5, 6 → total 4 columns
print("Cropped 4x4:\n", cropped_image)

plt.imshow(cropped_image, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.title("Cropped Region")
plt.show() 
```

6. Edge Detection (Simple - Convolution Concept): 
 
* Implement a very basic horizontal edge detection filter. This can be done by applying a convolution kernel (e.g., [-1, 0, 1] horizontally) to each row. For simplicity, you can just calculate the difference between adjacent pixels image[i, j+1] - image[i, j] for each pixel. Handle boundaries carefully (e.g., by skipping the last column).

* Big Change (Edge) | 50 → 200|150 **Edge Detected** 
* Small Change (Smooth)| 128 → 130|2 **No Edge**       
* Same Value | 90 → 90 | 0 |**Flat Area**     


```python
#first orignal
image = np.random.randint(0, 256, size=(10, 10), dtype=np.uint8)
print("Original Image:\n", image)

#horizontal Edge Detection 
edges = np.zeros_like(image, dtype=int)

for i in range(image.shape[0]):           # Go through each row
    for j in range(image.shape[1] - 1):   # Go through each column except the last
        edges[i, j] = int(image[i, j+1]) - int(image[i, j])
print("horizontal edge:\n",edges)


# display image
plt.imshow(edges, cmap='bwr') #Blue-White-Red colormap. 
plt.colorbar()
plt.title("Horizontal Edges (Difference of Adjacent Pixels)")
plt.show()

#notes
#+ Positive values where brightness increases from left to right.

#- Negative values where brightness decreases from left to right.

# 0 when no change.
```
